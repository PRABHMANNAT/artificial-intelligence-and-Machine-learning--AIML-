## 📌 What is Seaborn?

**Seaborn** is a Python visualization library **built on top of Matplotlib** that provides a **higher-level, more beautiful** interface for creating statistical graphics.

### Seaborn vs Matplotlib:
| Feature | Matplotlib | Seaborn |
|---------|-----------|--------|
| Code length | More lines needed | Fewer lines |
| Default style | Basic/plain | Beautiful & polished |
| Statistical features | Manual | Built-in (estimators, CI) |
| Built-in datasets | None | Yes (`load_dataset`) |

> **Think of it this way:** Matplotlib gives you a blank canvas and paints. Seaborn gives you a pre-designed template.

## 🔵 SECTION 2: Scatter Plot

Shows the **relationship between two numerical variables**. Each dot = one record.


````python
sns.scatterplot(
    x='total_bill',   # X-axis column name
    y='tip',          # Y-axis column name
    data=tips         # Which dataset to use
)

plt.title('Scatter Plot: Total Bill vs Tip')
plt.show()
````

### 💡 Bonus — Add color by category:
```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
# hue → colors each point by category (e.g., Male/Female)
```

---

## 📈 SECTION 3: Line Plot

Shows **trend or change** — connects data points with a line. Automatically computes **mean + confidence interval** (shaded area).


````python
sns.lineplot(
    x='size',           # Group size (how many people)
    y='total_bill',     # Bill amount
    data=tips
)

plt.title('Line Plot: Total Bill by Group Size')
plt.show()
````

> 💡 The **shaded region** around the line is the **95% confidence interval** — Seaborn adds this automatically! Matplotlib doesn't do this.

---

## 📊 SECTION 4: Bar Plot (Categorical)

Compares a **numerical value across categories**. Seaborn automatically shows the **mean + confidence interval** per category.


````python
sns.barplot(
    x='day',            # Categorical x-axis
    y='total_bill',     # Numerical y-axis
    data=tips
)

plt.title('Bar Plot: Average Total Bill by Day')
plt.show()

# The black line on top of each bar = confidence interval (uncertainty range)
````

> 💡 By default `sns.barplot()` shows the **mean** of Y for each X category. The error bar = confidence interval.

---

## 📦 SECTION 5: Box Plot

Box plot shows the **5-number summary + outliers** for a numerical column, split by categories.

### Box Plot Anatomy:
```
            ┌─────────┐
  ──────────┤  50th % ├──────────    ← Median (middle line)
            │  box    │
  ──────────┘  25th % └──────────    ← IQR (box = Q1 to Q3)
     │                    │          ← Whiskers (min/max within 1.5×IQR)
     ○                    ○          ← Outliers (dots beyond whiskers)
```


````python
sns.boxplot(
    x='day',
    y='total_bill',
    data=tips
)

plt.title('Box Plot: Total Bill Distribution by Day')
plt.show()

# Dots above/below whiskers = OUTLIERS
````

| Part of Box | What it means |
|------------|---------------|
| Bottom whisker | Minimum (0th percentile) |
| Bottom of box | Q1 (25th percentile) |
| Middle line | Median (50th percentile) |
| Top of box | Q3 (75th percentile) |
| Top whisker | Maximum (100th percentile within range) |
| Dots beyond whiskers | Outliers |

---

## 🎻 SECTION 6: Violin Plot

Violin plot = **Box Plot + KDE (Kernel Density Estimate)**. It shows the **full distribution shape**, not just percentiles.

- **Wider part** = more data points in that value range
- **Narrow part** = fewer data points


````python
sns.violinplot(
    x='day',
    y='total_bill',
    data=tips
)

plt.title('Violin Plot: Total Bill Distribution by Day')
plt.show()
````

### Box Plot vs Violin Plot:
| | Box Plot | Violin Plot |
|---|---------|-------------|
| Shows | 5-number summary | Full distribution shape |
| Outliers | Visible as dots | Hidden inside shape |
| Best for | Comparing medians & spread | Comparing distribution shapes |

---

## 📉 SECTION 7: Histogram (histplot)

Shows **how values are distributed** across bins. Seaborn's `histplot` can also overlay a **KDE curve** (distribution shape).


````python
# With KDE curve overlaid
sns.histplot(
    tips['total_bill'],
    bins=10,
    kde=True       # True = draw smooth distribution curve on top
)

plt.title('Histogram: Distribution of Total Bill')
plt.show()
````

````python
# Without KDE (just bars)
sns.histplot(tips['total_bill'], bins=10, kde=False)
plt.title('Histogram without KDE')
plt.show()
````

> 💡 `kde=True` smoothens the histogram into a curve → shows the **distribution type** (normal, skewed, bimodal, etc.)

---

## 〰️ SECTION 8: KDE Plot (Kernel Density Estimate)

KDE plot shows **only the smooth distribution curve** — without the histogram bars. Best for understanding data shape.


````python
sns.kdeplot(
    tips['total_bill'],
    fill=True        # Fill area under curve (shade=True is deprecated, use fill=True)
)

plt.title('KDE Plot: Total Bill Distribution')
plt.show()
````

### Histogram vs KDE:
| | Histogram | KDE Plot |
|---|-----------|----------|
| Shows | Counts per bin | Smooth density curve |
| Good for | Counting frequency | Understanding shape |
| Combined | `histplot(kde=True)` | Best of both worlds |

## 🔢 SECTION 9: Pair Plot

Pair plot shows the **relationship between ALL pairs of numerical variables** in a dataset — in one grid!

- **Off-diagonal cells** = Scatter plot of feature A vs feature B
- **Diagonal cells** = Histogram/KDE of that feature with itself


````python
# Creates an n×n grid for all numerical columns
sns.pairplot(tips)
plt.suptitle('Pair Plot: All Numerical Feature Relationships', y=1.02)
plt.show()

# Note: Categorical columns (sex, day, smoker) are automatically ignored
````

````python
# Bonus: Color by category using hue
sns.pairplot(tips, hue='sex')   # Male vs Female colored differently
plt.show()
````

> 💡 **Pair plot is your first stop in EDA** — one command reveals all relationships between features! Only works with numerical variables.

---

## 🌡️ SECTION 10: Heatmap (Correlation Matrix)

Heatmap shows **correlation between all numerical features** using a colored grid.

### What is Correlation?
- Value between **-1 and +1**
- **+1** = Perfect positive correlation (A increases → B increases)
- **0** = No relationship
- **-1** = Perfect negative correlation (A increases → B decreases)


````python
# Step 1: Calculate correlation matrix (only numerical columns)
corr_matrix = tips[['total_bill', 'tip', 'size']].corr()
print(corr_matrix)
````

````python
# Step 2: Visualize correlation as heatmap
sns.heatmap(
    corr_matrix,
    annot=True,           # Show numbers inside each cell
    cmap='coolwarm'       # Color scheme: red=positive, blue=negative
)

plt.title('Correlation Heatmap: Tips Dataset')
plt.show()

# Red = high positive correlation
# Blue = high negative correlation
# White/light = near zero (no relationship)
````

### 💡 Reading the Heatmap:
- `total_bill ↔ tip = 0.68` → Strong positive: bigger bill = bigger tip
- `total_bill ↔ size = 0.60` → Moderate positive: bigger group = bigger bill
- Diagonal is always **1.0** (a feature is perfectly correlated with itself)

## 🌍 SECTION 11: Real Sales Data Visualization

Combine **Pandas + Seaborn** on real CSV data.


````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load sales data
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
````

````python
# ── Chart 1: Total Revenue by Product Category ──
plt.figure(figsize=(10, 6))

sns.barplot(
    x='Product Category',
    y='Total Revenue',
    data=sales_df,
    estimator=sum         # Use SUM as aggregation (default is mean)
)

plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
# ── Chart 2: Total Revenue by Region ──
plt.figure(figsize=(10, 6))

sns.barplot(
    x='Region',
    y='Total Revenue',
    data=sales_df,
    estimator=sum
)

plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()

# Insight: North America > Asia > Europe (from the data)
````

### 💡 Key Seaborn Advantage:
In **Matplotlib**, you had to manually `groupby().sum()` before plotting.
In **Seaborn**, just pass `estimator=sum` and it handles aggregation **automatically**!

# 🚀 Seaborn Cheatsheet & Summary Sheet


```python
import seaborn as sns
import matplotlib.pyplot as plt

# ── LOAD DATASET ─────────────────────────────────────────
df = sns.load_dataset('tips')         # Built-in dataset

# ── RELATIONAL PLOTS ─────────────────────────────────────
sns.scatterplot(x='col1', y='col2', data=df, hue='cat_col')
sns.lineplot(x='col1', y='col2', data=df)

# ── CATEGORICAL PLOTS ────────────────────────────────────
sns.barplot(x='cat', y='num', data=df, estimator=sum)
sns.boxplot(x='cat', y='num', data=df)
sns.violinplot(x='cat', y='num', data=df)

# ── DISTRIBUTION PLOTS ───────────────────────────────────
sns.histplot(df['col'], bins=10, kde=True)
sns.kdeplot(df['col'], fill=True)

# ── MULTI-FEATURE PLOTS ──────────────────────────────────
sns.pairplot(df)                      # All feature pairs
sns.pairplot(df, hue='cat_col')       # Color by category

# ── HEATMAP ──────────────────────────────────────────────
corr = df[['num1','num2','num3']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')

# ── CUSTOMIZATION ────────────────────────────────────────
plt.figure(figsize=(10, 6))           # Set figure size
plt.title('My Chart')                 # Title
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.xticks(rotation=45)               # Rotate x labels
plt.tight_layout()                    # Fix spacing
plt.show()                            # Display
plt.savefig('chart.png', dpi=150)    # Save
```

# ❓ Important Questions & Answers


**Q1. Seaborn kya hai aur Matplotlib se kya fark hai?**
> Seaborn ek high-level visualization library hai jo Matplotlib ke upar bani hai. Matplotlib mein zyada code likhna padta hai aur simple style milti hai. Seaborn mein kam code mein statistical plots (CI, KDE) automatically aate hain aur style better hoti hai.

**Q2. Seaborn ka standard alias kya hai?**
> `import seaborn as sns` — `sns` standard alias hai.

**Q3. `sns.load_dataset()` kya karta hai?**
> Seaborn ke built-in practice datasets load karta hai (jaise 'tips', 'iris', 'titanic') — internet se bina download kiye directly use kar sakte ho.

**Q4. Seaborn ke barplot mein confidence interval kya hota hai?**
> Har bar ke upar ki kaali line = 95% confidence interval. Matlab: agar experiment dobara karo, toh actual mean is range mein 95% chance se hogi.

**Q5. Box plot ke parts kaunse hain?**
> Bottom whisker (min), Q1 box bottom (25%), median line (50%), Q3 box top (75%), top whisker (max within 1.5×IQR), dots = outliers.

**Q6. Violin Plot aur Box Plot mein kya fark hai?**
> Box plot = percentiles + outlier dots. Violin plot = box plot + full distribution shape (KDE curve). Violin zyada information deta hai.

**Q7. `sns.histplot()` mein `kde=True` kya karta hai?**
> Histogram bars ke upar ek smooth distribution curve draw karta hai — data kis distribution ko follow karta hai yeh clearly dikhta hai.

**Q8. KDE Plot kya hota hai?**
> Kernel Density Estimate — ek smooth curve jo data ki probability distribution dikhata hai. Histogram ka smoother version.

**Q9. `fill=True` KDE mein kyon use karte hain?**
> Curve ke neeche area fill/shade karta hai — visually better lagta hai. (Purana `shade=True` deprecated ho gaya, ab `fill=True` use karo.)

**Q10. Pair Plot kab use karte hain?**
> Jab ek saath saare numerical features ke beech relationships dekhne ho — EDA ka pehla step. Ek command mein n×n grid ban jaata hai.

**Q11. Pair Plot mein diagonal mein kya hota hai?**
> Har feature ka apna histogram ya KDE plot — kyunki feature se feature ki scatter plot boring hogi (straight line).

**Q12. Heatmap mein `annot=True` kya karta hai?**
> Har cell ke andar actual correlation value number print karta hai.

**Q13. Correlation ki value interpret kaise karein?**
> +1 = perfect positive, 0 = no relation, -1 = perfect negative. 0.5-0.7 = moderate positive, 0.7+ = strong positive.

**Q14. Seaborn ke `estimator=sum` ka matlab?**
> `sns.barplot()` by default mean dikhata hai. `estimator=sum` pass karo toh total sum dikhata hai — aise groupby manually nahi karna padta!

**Q15. `hue` parameter kya karta hai?**
> Ek categorical column pass karo — us column ki values ke basis pe har data point/bar alag color mein dikhega. Jaise `hue='sex'` → Male alag color, Female alag color.
