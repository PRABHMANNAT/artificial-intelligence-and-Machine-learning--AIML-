## 📊 SECTION 1 — Pandas Series

### What is a Series?
- Series ek **1D (one-dimensional)** array-like object hai.
- Kisi bhi data type ke elements hold kar sakti hai (int, float, string, etc.).
- Ek **table ke single column** ki tarah sochlo.
- Har element ka ek **index** hota hai (default: 0, 1, 2, ...)

```
 Index | Value
-------+------
   0   |  12
   1   |  34
   2   |  56
```

### 3 Ways to Create a Series:
1. **From a List** — default numeric index milega
2. **From a Dictionary** — keys automatically index ban jaati hain
3. **Custom Index** — khud index define karo


````python
# ── Way 1: Series from a List ──
data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print('Series from list:')
print(series)
print('\nData type:', type(series))   # pandas.core.series.Series
````

````python
# ── Way 2: Series from a Dictionary ──
# Dictionary ki KEYS → index ban jaati hain
# Dictionary ki VALUES → series values ban jaati hain
data_dict = {'a': 1, 'b': 2, 'c': 3}
series_dict = pd.Series(data_dict)
print('Series from dictionary:')
print(series_dict)
# Index: a, b, c  (keys)
# Values: 1, 2, 3 (values)
````

````python
# ── Way 3: Series with Custom Index ──
# Apna khud ka index define karo
data = [10, 20, 30]
indexes = ['a', 'b', 'c']   # number of indexes == number of elements

custom_series = pd.Series(data, index=indexes)
print('Series with custom index:')
print(custom_series)
````

## 🗂️ SECTION 2 — Pandas DataFrame

### What is a DataFrame?
- DataFrame ek **2D (two-dimensional)** data structure hai.
- **Mutable** (changes ho sakti hain) aur **heterogeneous** (different types of data ho sakta hai).
- Ek **Excel sheet ya database table** ki tarah sochlo.
- Rows + Columns dono hote hain — **labeled axes**.

```
 Index | Name   | Age | City
-------+--------+-----+-----------
   0   | Krish  |  25 | Bangalore
   1   | John   |  30 | New York
   2   | Jack   |  45 | Florida
```

### Series vs DataFrame:
| Feature       | Series        | DataFrame           |
|---------------|---------------|---------------------|
| Dimensions    | 1D            | 2D                  |
| Structure     | Single column | Multiple rows & cols|
| Analogy       | One column    | Full Excel Sheet    |

### 3 Ways to Create a DataFrame:
1. **From Dictionary of Lists** — most common
2. **From List of Dictionaries** — row-wise data
3. **From CSV/Excel file** — real-world datasets


````python
# ── Way 1: DataFrame from Dictionary of Lists ──
# Key = column name, Value = list of column data
data = {
    'Name': ['Krish', 'John', 'Jack'],
    'Age':  [25, 30, 45],
    'City': ['Bangalore', 'New York', 'Florida']
}
df = pd.DataFrame(data)
print('DataFrame from dictionary of lists:')
print(df)
print('\nData type:', type(df))
````

````python
# ── Way 2: DataFrame from List of Dictionaries ──
# Har dictionary = ek ROW of data
data = [
    {'Name': 'Krish', 'Age': 32, 'City': 'Bangalore'},
    {'Name': 'John',  'Age': 34, 'City': 'Bangalore'},
    {'Name': 'Jack',  'Age': 32, 'City': 'Bangalore'}
]
df2 = pd.DataFrame(data)
print('DataFrame from list of dictionaries:')
print(df2)
````

## 🔍 SECTION 3 — Accessing Data from DataFrame

### 5 Ways to Access Data:

| Method          | Use For                          | Example                      |
|-----------------|----------------------------------|------------------------------|
| `df['col']`     | Single column (returns Series)   | `df['Name']`                 |
| `df.loc[r]`     | Row by **label/index name**      | `df.loc[0]`                  |
| `df.iloc[r, c]` | Row+Col by **integer position**  | `df.iloc[0, 2]`              |
| `df.at[r, col]` | Single value by label            | `df.at[1, 'Age']`            |
| `df.iat[r, c]`  | Single value by integer position | `df.iat[2, 2]`               |

> 💡 **Key Rule:** Ek column pick karo → Series milegi. Poora DataFrame → DataFrame milega.


````python
# ── Method 1: Access column by name ──
print('All names (df["Name"]):')
print(df['Name'])                        # returns a Series
print('Type:', type(df['Name']))         # pandas.core.series.Series
````

````python
# ── Method 2: loc — access by row label (index) ──
print('Row at index 0 (df.loc[0]):')
print(df.loc[0])          # zeroth row ki poori info

print('\nRows 0 to 1 (df.loc[0:1]):')
print(df.loc[0:1])        # inclusive slice
````

````python
# ── Method 3: iloc — access by integer position ──
print('Element at row 0, col 0 (df.iloc[0, 0]):', df.iloc[0, 0])   # Krish
print('Element at row 0, col 1 (df.iloc[0, 1]):', df.iloc[0, 1])   # 25
print('Element at row 2, col 2 (df.iloc[2, 2]):', df.iloc[2, 2])   # Florida
````

````python
# ── Method 4: at — single value by row label + column name ──
# df.at[row_index, 'column_name']
print('df.at[1, "Age"]   →', df.at[1, 'Age'])    # 30
print('df.at[2, "Name"]  →', df.at[2, 'Name'])   # Jack
print('df.at[0, "City"]  →', df.at[0, 'City'])   # Bangalore
````

````python
# ── Method 5: iat — single value by row + col INTEGER position ──
# df.iat[row_pos, col_pos]
print('df.iat[2, 2]  →', df.iat[2, 2])   # Florida (row=2, col=2)
print('df.iat[0, 0]  →', df.iat[0, 0])   # Krish
print('df.iat[1, 1]  →', df.iat[1, 1])   # 30
````

## ✏️ SECTION 4 — Data Manipulation with DataFrames

### Operations Covered:
1. **Add a new column**
2. **Modify existing column values**
3. **Drop a column** (axis=1)
4. **Drop a row** (axis=0)
5. **inplace=True** — permanent vs temporary operations

### ⚠️ Important: `inplace` Parameter
- By default, `drop()` ek **temporary** operation karta hai — original df change nahi hota.
- `inplace=True` lagao to change **permanent** ho jaata hai.
```python
df.drop('Salary', axis=1, inplace=True)  # PERMANENT
```

### `axis` Parameter:
| axis value | Matlab                          |
|------------|----------------------------------|
| `axis=0`   | Default — rows pe operation      |
| `axis=1`   | Columns pe operation             |


````python
# Fresh DataFrame
df = pd.DataFrame({
    'Name': ['Krish', 'John', 'Jack'],
    'Age':  [25, 30, 45],
    'City': ['Bangalore', 'New York', 'Florida']
})

# ── Operation 1: Add a new column ──
df['Salary'] = [50000, 60000, 70000]
print('After adding Salary column:')
print(df)
````

````python
# ── Operation 2: Modify column values ──
# Sabke age mein 1 add karo (next year)
df['Age'] = df['Age'] + 1
print('After incrementing Age by 1:')
print(df)
# 25→26, 30→31, 45→46
````

````python
# ── Operation 3: Drop a column (axis=1) ──
# axis=1 means column  |  inplace=True means permanent
df.drop('Salary', axis=1, inplace=True)
print('After dropping Salary column:')
print(df)
````

````python
# ── Operation 4: Drop a row (axis=0) ──
# Row index 0 (Krish) drop karo
df.drop(0, axis=0, inplace=True)
print('After dropping row at index 0 (Krish):')
print(df)
# Only John and Jack remain
````

## 📋 SECTION 5 — Useful DataFrame Attributes & Methods

| Method/Attribute   | Kya karta hai                                           |
|--------------------|----------------------------------------------------------|
| `df.head(n)`       | Top n rows dikhata hai (default=5)                       |
| `df.tail(n)`       | Last n rows dikhata hai (default=5)                      |
| `df.describe()`    | Statistical summary — mean, std, min, max, percentiles   |
| `df.dtypes`        | Har column ka data type                                  |
| `df.shape`         | (rows, columns) — tuple                                  |
| `df.columns`       | Column names list                                        |
| `df.index`         | Row index info                                           |
| `df.info()`        | DataFrame ka overview — nulls, dtypes, memory            |
| `df.isnull()`      | Check for missing values (NaN)                           |


````python
# Demo DataFrame for attributes
df_demo = pd.DataFrame({
    'Name':   ['Krish', 'John', 'Jack', 'Sara', 'Tom'],
    'Age':    [25, 30, 45, 28, 35],
    'Salary': [50000, 60000, 70000, 55000, 65000],
    'City':   ['Bangalore', 'New York', 'Florida', 'Chicago', 'Texas']
})

print('─── df.head(3) ───')
print(df_demo.head(3))

print('\n─── df.tail(2) ───')
print(df_demo.tail(2))
````

````python
print('─── df.describe() — Statistical Summary ───')
print(df_demo.describe())
# count, mean, std, min, 25%, 50%, 75%, max
````

````python
print('─── df.dtypes ───')
print(df_demo.dtypes)

print('\n─── df.shape ───')
print(df_demo.shape)        # (5, 4) → 5 rows, 4 columns

print('\n─── df.columns ───')
print(df_demo.columns)

print('\n─── df.info() ───')
df_demo.info()
````

## 🔄 SECTION 6 — NumPy ↔ Pandas Conversion

- Pandas DataFrames ko NumPy arrays mein convert kiya ja sakta hai.
- `np.array(df)` se DataFrame → NumPy array (index/columns skip ho jaate hain).


````python
# DataFrame → NumPy Array
df_nums = pd.DataFrame({'A': [1,2,3], 'B': [4,5,6], 'C': [7,8,9]})
arr = np.array(df_nums)
print('DataFrame to NumPy array:')
print(arr)
print('Type:', type(arr))

# NumPy Array → DataFrame
df_from_arr = pd.DataFrame(arr, columns=['A', 'B', 'C'])
print('\nNumPy array to DataFrame:')
print(df_from_arr)
````


---
## 🔍 Accessing Data
```python
df['column']              # single column → Series
df[['col1','col2']]       # multiple columns → DataFrame
df.loc[0]                 # row by label
df.loc[0:2]               # rows 0 to 2 (inclusive)
df.iloc[0, 1]             # row 0, col 1 (integer pos)
df.at[1, 'Age']           # single value by label
df.iat[2, 2]              # single value by position
```

---
## ✏️ Data Manipulation
```python
df['NewCol'] = [v1, v2, v3]             # Add column
df['Age'] = df['Age'] + 1               # Modify column
df.drop('ColName', axis=1, inplace=True)  # Drop column (permanent)
df.drop(0, axis=0, inplace=True)          # Drop row by index (permanent)
# Without inplace=True → temporary change only!
```

---
## 📋 Useful Methods
```python
df.head(n)       # first n rows
df.tail(n)       # last n rows
df.describe()    # stats: count, mean, std, min, max, percentiles
df.info()        # dtypes, nulls, memory
df.dtypes        # data type of each column
df.shape         # (rows, cols)
df.columns       # column names
df.index         # row index
df.isnull()      # check for NaN values

```
