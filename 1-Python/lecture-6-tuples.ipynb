# 1. Introduction to Tuples

A **tuple** is one of the built-in data structures in Python.

## Definition
A tuple is an **ordered, immutable collection of items**.

### What this means
- **ordered** → items keep their position
- **immutable** → items cannot be changed after creation
- **collection** → a tuple can hold multiple values together

## Similarity with lists
Tuples are similar to lists because:
- both are ordered
- both can contain multiple items
- both support indexing and slicing
- both can hold mixed data types

## Main difference from lists
- **list** → mutable
- **tuple** → immutable


## Why tuples are useful
Tuples are useful when:
- data should not change
- you want fixed records
- values need to be safely grouped
- returning multiple values from functions
- using data as dictionary keys (when valid)


# 2. Creating Tuples

## Empty tuple
Use parentheses:
```python
empty_tuple = ()
```


````python
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
````

## Create tuple using `tuple()`
You can also use the built-in `tuple()` constructor.


````python
empty_tuple_2 = tuple()
print(empty_tuple_2)
print(type(empty_tuple_2))
````

## Empty list for comparison


````python
empty_list = []
print(empty_list)
print(type(empty_list))

empty_list_2 = list()
print(empty_list_2)
print(type(empty_list_2))
````

## Tuple with multiple values


````python
numbers = (1, 2, 3, 4, 5, 6)
print(numbers)
print(type(numbers))
````

## Tuple without parentheses
Commas are enough in many cases:


````python
numbers2 = 1, 2, 3
print(numbers2)
print(type(numbers2))
````

## Single-element tuple
This is a very important special case.


````python
single_tuple = (5,)
print(single_tuple)
print(type(single_tuple))

not_a_tuple = (5)
print(not_a_tuple)
print(type(not_a_tuple))
````

### Explanation
- `(5,)` is a tuple
- `(5)` is just an integer inside parentheses


## Mixed-type tuple
A tuple can contain different data types.


````python
mixed_tuple = (1, "hello", 3.14, True)
print(mixed_tuple)
print(type(mixed_tuple))
````

# 3. Converting Between List and Tuple

A list can be converted to a tuple.


````python
num_list = [1, 2, 3, 4, 5, 6]
num_tuple = tuple(num_list)

print(num_list)
print(num_tuple)
print(type(num_tuple))
````

A tuple can also be converted to a list.


````python
numbers = (1, 2, 3, 4, 5, 6)
numbers_list = list(numbers)

print(numbers)
print(numbers_list)
print(type(numbers_list))
````

# 4. Accessing Tuple Elements

Tuples support indexing just like lists.

## Important rule
Python indexing starts at `0`.


````python
numbers = (1, 2, 3, 4, 5, 6)

print(numbers[0])   # first element
print(numbers[1])   # second element
print(numbers[-1])  # last element
print(numbers[-2])  # second last element
````

# 5. Slicing Tuples

Tuples support slicing just like lists.

## Syntax
```python
tuple[start:stop:step]
```

- start included
- stop excluded


````python
numbers = (1, 2, 3, 4, 5, 6)

print(numbers[0:4])
print(numbers[:4])
print(numbers[2:])
print(numbers[::2])
print(numbers[::-1])
````

## Explanation
- `numbers[0:4]` → first four elements
- `numbers[:4]` → same as above
- `numbers[2:]` → from index 2 to end
- `numbers[::2]` → every second element
- `numbers[::-1]` → reversed tuple


## Important note
When you slice a tuple, the result is also a **tuple**.


````python
numbers = (1, 2, 3, 4, 5, 6)
result = numbers[1:4]
print(result)
print(type(result))
````

# 6. Tuple Operations

Tuples support several common operations such as:
- concatenation
- repetition
- membership testing
- length checking


## 6.1 Concatenation
Use `+` to join two tuples.


````python
numbers = (1, 2, 3, 4, 5, 6)
mixed_tuple = (1, "hello", 3.14, True)

concatenated = numbers + mixed_tuple
print(concatenated)
````

## 6.2 Repetition
Use `*` to repeat a tuple.


````python
mixed_tuple = (1, "hello", 3.14, True)
repeated = mixed_tuple * 3
print(repeated)
````

### Important note
Tuple repetition repeats the **whole tuple**.  
It does not multiply numeric elements individually.


````python
numbers = (1, 2, 3)
print(numbers * 3)
````

## 6.3 Membership testing
Use `in` and `not in`.


````python
numbers = (1, 2, 3, 4, 5, 6)

print(3 in numbers)
print(10 in numbers)
print(10 not in numbers)
````

## 6.4 Length
Use `len()`.


````python
numbers = (1, 2, 3, 4, 5, 6)
print(len(numbers))
````

# 7. Immutable Nature of Tuples

This is the most important property of tuples.

## What does immutable mean?
It means once a tuple is created, its items cannot be changed directly.


## Compare with a list
Lists are mutable:


````python
my_list = [1, 2, 3, 4, 5]
my_list[1] = "Krish"
print(my_list)
````

## Tuples are immutable
Trying to modify an item raises an error.


````python
numbers = (1, 2, 3, 4, 5, 6)

# Uncomment to test immutability:
# numbers[1] = "Krish"
````

### Error you would get
```python
TypeError: 'tuple' object does not support item assignment
```


## How to "modify" a tuple in practice
You cannot change it directly, but you can:

1. convert tuple to list  
2. modify the list  
3. convert it back to tuple


````python
numbers = (1, 2, 3, 4, 5, 6)

temp_list = list(numbers)
temp_list[1] = "Krish"
numbers = tuple(temp_list)

print(numbers)
print(type(numbers))
````

# 8. Common Tuple Methods

Tuples have fewer methods than lists because tuples are immutable.

The two most common tuple methods are:
- `count()`
- `index()`


## 8.1 `count()`
Counts how many times a value appears.


````python
numbers = (1, 2, 3, 4, 5, 6, 1, 1)
print(numbers.count(1))
print(numbers.count(3))
````

## 8.2 `index()`
Returns the index of the first occurrence of a value.


````python
numbers = (1, 2, 3, 4, 5, 6)
print(numbers.index(3))
print(numbers.index(6))
````

### Important note
If the item is not present, `index()` raises `ValueError`.


````python
numbers = (1, 2, 3)

# Uncomment to test:
# print(numbers.index(10))
````

# 9. Packing and Unpacking Tuples

## Packing
Packing means grouping values together into a tuple.


````python
packed_tuple = 1, "hello", 3.14
print(packed_tuple)
print(type(packed_tuple))
````

## Unpacking
Unpacking means assigning tuple values to separate variables.


````python
packed_tuple = (1, "hello", 3.14)

a, b, c = packed_tuple
print(a)
print(b)
print(c)
````

### Important rule
The number of variables must match the number of tuple items, unless you use starred unpacking.


# 10. Unpacking with `*`

Starred unpacking lets one variable collect multiple items.


````python
numbers = (1, 2, 3, 4, 5, 6)

first, *middle, last = numbers

print("first =", first)
print("middle =", middle)
print("last =", last)
````

## Explanation
- `first` gets the first item
- `last` gets the last item
- `*middle` collects everything in between as a list


````python
numbers = (1, 2, 3, 4, 5, 6)

*beginning, last = numbers
print(beginning)
print(last)

first, *rest = numbers
print(first)
print(rest)
````

### Important note
The starred target receives a **list**, not a tuple.


````python
numbers = (1, 2, 3, 4)

first, *middle, last = numbers
print(type(middle))
````

# 11. Nested Tuples

A nested tuple is a tuple containing other tuples.


````python
nested_tuple = (
    (1, 2, 3),
    ("A", "B", "C"),
    (True, False)
)

print(nested_tuple)
````

## Accessing nested tuple elements


````python
nested_tuple = (
    (1, 2, 3),
    ("A", "B", "C"),
    (True, False)
)

print(nested_tuple[0])      # first inner tuple
print(nested_tuple[1][2])   # 'C'
print(nested_tuple[2][0])   # True
````

## Slicing nested tuple elements


````python
nested_tuple = (
    (1, 2, 3, 4),
    ("A", "B", "C", "D"),
    (True, False, True)
)

print(nested_tuple[0][0:3])
print(nested_tuple[1][1:])
````

# 12. Mixed Nested Structures

You can also have lists inside tuples or tuples inside lists.


````python
mixed_nested = (
    [1, 2, 3],
    ("hello", 3.14, True),
    ["A", "B", "C"]
)

print(mixed_nested)
print(mixed_nested[0])
print(mixed_nested[1][1])
````

### Important subtle point
A tuple is immutable, but if it contains a mutable object like a list, the list itself can still be changed.


````python
mixed_nested = (
    [1, 2, 3],
    ("hello", 3.14, True)
)

mixed_nested[0][1] = 999
print(mixed_nested)
````

# 13. Iterating Through Nested Tuples

You can loop through nested tuples using nested `for` loops.


````python
nested_tuple = (
    (1, 2, 3),
    ("A", "B", "C"),
    (True, False)
)

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end=" ")
    print()
````

## Explanation
- outer loop picks each inner tuple
- inner loop picks each item inside that tuple

# 16. Best Practices

## 1. Use tuples for fixed data
If values should not change, tuples are a good choice.

## 2. Use lists for changing collections
If you need frequent insertion, deletion, or modification, lists are usually better.

## 3. Avoid naming variables `tuple`
That hides the built-in `tuple` type.

## 4. Remember single-element tuple syntax
Always use the comma.

## 5. Use unpacking for cleaner code
Tuple unpacking often makes code more readable.

## 6. Use tuples as dictionary keys only when contents are hashable
For example, tuples containing lists cannot be used as dictionary keys.

# 20. Interview Concepts and Model Answers

### 1. What is a tuple in Python?
**Model answer:**  
A tuple is an ordered, immutable collection used to group related items together. It supports indexing and slicing and can store mixed data types.

### 2. What is the difference between a tuple and a list?
**Model answer:**  
Both are ordered collections, but lists are mutable while tuples are immutable. Lists use square brackets and tuples use parentheses.

### 3. Why would you use a tuple instead of a list?
**Model answer:**  
Use a tuple when the data should not change, when you want safer fixed records, or when you need hashable grouped values for use as dictionary keys.

### 4. What are tuple packing and unpacking?
**Model answer:**  
Packing means grouping multiple values into a tuple. Unpacking means assigning tuple elements into separate variables.

### 5. What is starred unpacking?
**Model answer:**  
Starred unpacking allows one variable to collect multiple remaining elements during unpacking, such as `first, *middle, last = values`.

### 6. What methods are commonly used with tuples?
**Model answer:**  
The main tuple methods are `count()` and `index()`.

### 7. Can a tuple be modified?
**Model answer:**  
Not directly. Tuples are immutable. To effectively modify one, convert it to a list, change it, and convert it back if needed.

### 8. What is a nested tuple?
**Model answer:**  
A nested tuple is a tuple that contains other tuples as elements.

### 9. Are tuples more memory-efficient than lists?
**Model answer:**  
In general, tuples are usually lighter than lists for fixed data because they are immutable, though exact memory behavior depends on Python implementation.

### 10. Can tuples be used as dictionary keys?
**Model answer:**  
Yes, but only if all their elements are hashable.
