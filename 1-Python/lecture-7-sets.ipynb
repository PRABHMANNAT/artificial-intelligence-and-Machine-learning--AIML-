# 1. Introduction to Sets

A **set** is a built-in Python data structure used to store a collection of **unique** items.

## Definition
A set is:
- **unordered**
- **mutable**
- contains **unique elements only**

## Key properties
- no duplicate elements
- order is not guaranteed
- useful for fast membership testing
- useful for removing duplicates
- useful for mathematical set operations


## Why sets are important
Sets are especially useful for:
- checking whether an element exists
- removing duplicates from data
- comparing groups of items
- finding common elements
- finding unique elements
- performing union, intersection, difference, and symmetric difference


# 2. Creating Sets

## Non-empty set
Use curly braces:
```python
my_set = {1, 2, 3, 4, 5}
```


````python
my_set = {1, 2, 3, 4, 5}
print(my_set)
print(type(my_set))
````

## Empty set
Use `set()`, not `{}`.


````python
empty_set = set()
print(empty_set)
print(type(empty_set))
````

## Why not `{}`?
Because:
```python
{}
```
creates an empty dictionary, not a set.


````python
not_a_set = {}
print(not_a_set)
print(type(not_a_set))
````

## Creating a set from a list
You can convert a list into a set using `set()`.


````python
numbers = [1, 2, 3, 4, 5, 6]
numbers_set = set(numbers)

print(numbers)
print(numbers_set)
print(type(numbers_set))
````

## Creating a set from a tuple


````python
values = (1, 2, 3, 4, 5, 6)
values_set = set(values)

print(values)
print(values_set)
print(type(values_set))
````

# 3. Unique Elements in Sets

Sets automatically remove duplicate elements.


````python
my_set = {1, 2, 3, 4, 5, 6, 5, 6, 4, 5}
print(my_set)
````

## Explanation
Even though duplicate values were written multiple times, the set keeps only one copy of each value.


# 4. Unordered Nature of Sets

Sets are unordered.

This means:
- elements do not have a fixed index
- you cannot access a set using `my_set[0]`
- display order may vary


````python
my_set = {"apple", "banana", "cherry", "kiwi"}
print(my_set)
````

### Important note
Do not assume the printed order is the insertion order.  
If you need ordered access, use a list or tuple instead.


# 5. Adding and Removing Elements

## 5.1 `add()`
Adds one element to a set.


````python
my_set = {1, 2, 3, 4, 5, 6}
my_set.add(7)

print(my_set)
````

### What if the element already exists?
Nothing changes, because sets only keep unique values.


````python
my_set = {1, 2, 3, 4, 5, 6}
my_set.add(6)

print(my_set)
````

## 5.2 `remove()`
Removes a specific element.

If the element is not present, Python raises `KeyError`.


````python
my_set = {1, 2, 3, 4, 5, 6}
my_set.remove(3)

print(my_set)
````

````python
my_set = {1, 2, 3, 4, 5, 6}

# Uncomment to test:
# my_set.remove(10)
````

## 5.3 `discard()`
Removes an element if it exists.

If it does not exist, no error is raised.


````python
my_set = {1, 2, 3, 4, 5, 6}
my_set.discard(10)

print(my_set)
````

### Difference between `remove()` and `discard()`
- `remove(x)` → raises error if `x` is missing
- `discard(x)` → silently does nothing if `x` is missing


## 5.4 `pop()`
Removes and returns an arbitrary element from the set.


````python
my_set = {1, 2, 3, 4, 5, 6}
removed_element = my_set.pop()

print("Removed:", removed_element)
print("Remaining set:", my_set)
````

### Important correction
Because sets are unordered, `pop()` does **not** mean first-in-first-out.
It removes an arbitrary element.


## 5.5 `clear()`
Removes all elements from the set.


````python
my_set = {1, 2, 3, 4, 5, 6}
my_set.clear()

print(my_set)
````

# 6. Membership Testing

Membership testing checks whether an element is present in a set.

Use:
- `in`
- `not in`


````python
my_set = {1, 2, 3, 4, 5}

print(3 in my_set)
print(10 in my_set)
print(10 not in my_set)
````

## Why sets are fast for membership tests
Sets are designed for efficient membership checking, which is one reason they are so commonly used.


# 7. Mathematical Set Operations

Set operations are one of the main reasons sets are powerful.

Let:
```python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}
```


````python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

print("set1 =", set1)
print("set2 =", set2)
````

## 7.1 Union
Union means all unique elements from both sets combined.

Method:
```python
set1.union(set2)
```

Operator:
```python
set1 | set2
```


````python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

union_set = set1.union(set2)
print(union_set)

print(set1 | set2)
````

## 7.2 Intersection
Intersection means only the common elements present in both sets.

Method:
```python
set1.intersection(set2)
```

Operator:
```python
set1 & set2
```


````python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

intersection_set = set1.intersection(set2)
print(intersection_set)

print(set1 & set2)
````

## 7.3 `intersection_update()`
This updates the original set to keep only common elements.


````python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

set1.intersection_update(set2)
print(set1)
````

### Important note
Methods ending with `_update()` usually change the original set in-place.


## 7.4 Difference
Difference means elements present in one set but not in the other.

Method:
```python
set1.difference(set2)
```

Operator:
```python
set1 - set2
```


````python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

difference_set = set1.difference(set2)
print(difference_set)

print(set1 - set2)
````

## Difference in the other direction
`set2 - set1` is different from `set1 - set2`.


````python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

print(set2.difference(set1))
print(set2 - set1)
````

## 7.5 `difference_update()`
This removes from the original set all elements found in another set.


````python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

set1.difference_update(set2)
print(set1)
````

## 7.6 Symmetric Difference
Symmetric difference means elements that are in either set, but **not in both**.

Method:
```python
set1.symmetric_difference(set2)
```

Operator:
```python
set1 ^ set2
```


````python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

sym_diff = set1.symmetric_difference(set2)
print(sym_diff)

print(set1 ^ set2)
````

## Explanation
Common elements (`4, 5, 6`) are removed.
Remaining unique elements from both sets are kept.


# 8. Subset and Superset Methods

## 8.1 `issubset()`
Checks whether all elements of one set are contained in another set.


````python
set1 = {1, 2, 3}
set2 = {1, 2, 3, 4, 5}

print(set1.issubset(set2))
print(set2.issubset(set1))
````

## 8.2 `issuperset()`
Checks whether a set contains all elements of another set.


````python
set1 = {1, 2, 3}
set2 = {1, 2, 3, 4, 5}

print(set2.issuperset(set1))
print(set1.issuperset(set2))
````

## Quick intuition
- subset → “Am I fully inside the other set?”
- superset → “Do I fully contain the other set?”

# 11. Practical Example: Count Unique Words in Text


````python
text = "In this tutorial we are discussing about sets in this tutorial"

words = text.split()
unique_words = set(words)

print("Words:", words)
print("Unique words:", unique_words)
print("Number of unique words:", len(unique_words))
````

## Explanation
- `split()` breaks the text into words
- converting to `set` removes duplicates
- `len()` gives the count of unique words

# 14. Best Practices

## 1. Use sets when uniqueness matters
If duplicate values should not exist, sets are a strong choice.

## 2. Use sets for fast membership checks
This is one of their biggest advantages.

## 3. Do not depend on order
If order matters, use a list or another ordered structure.

## 4. Be careful with `remove()`
Use `discard()` if you want safer deletion without errors.

- valid → integers, strings, tuples of hashable items
- invalid → lists, dictionaries, mutable sets

# 17. Tricky Questions

### 1. Can you access `my_set[0]`?
**Answer:**  
No. Sets are unordered and do not support indexing.

### 2. What happens if you add the same element twice?
**Answer:**  
The set still keeps only one copy.

### 3. Does `pop()` remove the first inserted element from a set?
**Answer:**  
No. It removes an arbitrary element.

### 4. Can a set contain a list?
**Answer:**  
No. Lists are unhashable.

### 5. Can a set contain a tuple?
**Answer:**  
Yes, if the tuple contains only hashable elements.

### 6. What is the result of:
```python

```
**Answer:**  
`{1, 2, 3}`

### 7. Is this valid?
```python
s = {{1, 2}, {3, 4}}
```
**Answer:**  
No. Regular sets cannot be inside another set because they are mutable and unhashable.

### 8. What is the difference between `difference()` and `symmetric_difference()`?
**Answer:**  
`difference()` keeps items from one set that are not in the other.  
`symmetric_difference()` keeps non-common items from both sets.
