# 2. Example 1: Temperature Conversion

We want a function that converts:
- Celsius → Fahrenheit
- Fahrenheit → Celsius


## Formula recap

### Celsius to Fahrenheit
```python
F = (C * 9/5) + 32
```

### Fahrenheit to Celsius
```python
C = (F - 32) * 5/9
```


````python
def convert_temperature(temperature, unit):
    """Convert temperature between Celsius and Fahrenheit."""
    if unit == "C":
        return (temperature * 9 / 5) + 32
    elif unit == "F":
        return (temperature - 32) * 5 / 9
    else:
        return None

print(convert_temperature(25, "C"))   # Celsius to Fahrenheit
print(convert_temperature(77, "F"))   # Fahrenheit to Celsius
print(convert_temperature(10, "X"))   # Invalid unit
````

## Explanation
- if the unit is `"C"`, the function assumes the input is Celsius and converts it to Fahrenheit
- if the unit is `"F"`, the function assumes the input is Fahrenheit and converts it to Celsius
- if the unit is something else, the function returns `None`


## Better version with clearer validation


````python
def convert_temperature_safe(temperature, unit):
    """Convert temperature between Celsius and Fahrenheit with validation."""
    unit = unit.upper()

    if unit == "C":
        return (temperature * 9 / 5) + 32
    if unit == "F":
        return (temperature - 32) * 5 / 9

    raise ValueError("Unit must be 'C' or 'F'.")

print(convert_temperature_safe(25, "c"))
print(convert_temperature_safe(77, "f"))
````

# 3. Example 2: Password Strength Checker

A strong password usually needs:
- minimum length
- uppercase letter
- lowercase letter
- digit
- special character


````python
def is_strong_password(password):
    """Check whether a password is strong."""
    if len(password) < 8:
        return False

    if not any(ch.isdigit() for ch in password):
        return False

    if not any(ch.islower() for ch in password):
        return False

    if not any(ch.isupper() for ch in password):
        return False

    special_chars = "!@#$%^&*()-_=+[]{}|;:'\",.<>/?`~"
    if not any(ch in special_chars for ch in password):
        return False

    return True

print(is_strong_password("AaPassword"))
print(is_strong_password("Str0ngPwd!"))
````

## Explanation
The function returns `False` as soon as one required condition fails.

It returns `True` only when all checks are satisfied:
- length
- digit
- lowercase
- uppercase
- special character


## Why `any()` is useful here
The `any()` function checks whether **at least one** value in an iterable is true.

Example:
```python
any(ch.isdigit() for ch in password)
```
means:
> Does the password contain at least one digit?


## Slightly more user-friendly version
Instead of just `True/False`, we can return messages.


````python
def password_strength_message(password):
    if len(password) < 8:
        return "Weak: Password must be at least 8 characters long."
    if not any(ch.isdigit() for ch in password):
        return "Weak: Password must contain at least one digit."
    if not any(ch.islower() for ch in password):
        return "Weak: Password must contain at least one lowercase letter."
    if not any(ch.isupper() for ch in password):
        return "Weak: Password must contain at least one uppercase letter."
    if not any(ch in "!@#$%^&*()-_=+[]{}|;:'\",.<>/?`~" for ch in password):
        return "Weak: Password must contain at least one special character."
    return "Strong password."

print(password_strength_message("abc"))
print(password_strength_message("Strong1!"))
````

# 4. Example 3: Calculate Total Cost of Items in a Shopping Cart

This example uses:
- a function
- a list
- dictionaries

Each item in the cart has:
- name
- price
- quantity


````python
def calculate_total_cost(cart):
    """Calculate total cost of items in a shopping cart."""
    total_cost = 0
    for item in cart:
        total_cost += item["price"] * item["quantity"]
    return total_cost

cart = [
    {"name": "Apple", "price": 0.5, "quantity": 4},
    {"name": "Banana", "price": 0.3, "quantity": 6},
    {"name": "Orange", "price": 0.7, "quantity": 3},
]

total_cost = calculate_total_cost(cart)
print(total_cost)
````

## Explanation
For each item in the cart:
- take the item price
- multiply by quantity
- keep adding to `total_cost`

This is a common real-world pattern when working with:
- carts
- invoices
- bills
- order summaries


## Expanded version with itemized bill


````python
def calculate_total_cost_verbose(cart):
    total_cost = 0
    for item in cart:
        item_total = item["price"] * item["quantity"]
        total_cost += item_total
        print(f"{item['name']}: {item['quantity']} x {item['price']} = {item_total}")
    print(f"Total cost = {total_cost}")
    return total_cost

cart = [
    {"name": "Apple", "price": 0.5, "quantity": 4},
    {"name": "Banana", "price": 0.3, "quantity": 6},
    {"name": "Orange", "price": 0.7, "quantity": 3},
]

calculate_total_cost_verbose(cart)
````

# 5. Example 4: Check if a String Is a Palindrome

A palindrome reads the same:
- from left to right
- from right to left

Examples:
- `madam`
- `racecar`
- `A man a plan a canal Panama` (if spaces/case are ignored)


````python
def is_palindrome(s):
    """Check whether a string is a palindrome."""
    s = s.lower().replace(" ", "")
    return s == s[::-1]

print(is_palindrome("A man a plan a canal Panama"))
print(is_palindrome("hello"))
````

## Explanation
- `lower()` makes comparison case-insensitive
- `replace(" ", "")` removes spaces
- `s[::-1]` reverses the string
- then the function compares the cleaned string with its reverse


## Improved version handling punctuation too


````python
import re

def is_palindrome_clean(s):
    cleaned = re.sub(r"[^a-zA-Z0-9]", "", s.lower())
    return cleaned == cleaned[::-1]

print(is_palindrome_clean("Madam, I'm Adam"))
print(is_palindrome_clean("Python"))
````

# 6. Example 5: Factorial Using Recursion

## What is factorial?
```python
5! = 5 × 4 × 3 × 2 × 1 = 120
```

## What is recursion?
Recursion means a function calls itself.


````python
def factorial(n):
    """Return factorial of a number using recursion."""
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
print(factorial(6))
````

## How recursion works for `factorial(5)`
```python
factorial(5)
= 5 * factorial(4)
= 5 * 4 * factorial(3)
= 5 * 4 * 3 * factorial(2)
= 5 * 4 * 3 * 2 * factorial(1)
= 5 * 4 * 3 * 2 * 1 * factorial(0)
= 5 * 4 * 3 * 2 * 1 * 1
= 120
```


## Safer version with validation


````python
def factorial_safe(n):
    if not isinstance(n, int):
        raise TypeError("Factorial is defined for integers only.")
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    if n == 0:
        return 1
    return n * factorial_safe(n - 1)

print(factorial_safe(5))
````

# 7. Example 6: Read a File and Count Word Frequency

This example combines:
- functions
- file handling
- dictionaries
- string cleanup


## Sample text file setup
The transcript used a sample text file, so this notebook first creates one.


````python
sample_text = """Hello world
How are you
My name is Krish
Krish likes Python
"""

with open("sample.txt", "w", encoding="utf-8") as file:
    file.write(sample_text)

print("sample.txt created")
````

````python
def count_word_frequency(file_path):
    """Read a file and count frequency of each word."""
    word_count = {}

    with open(file_path, "r", encoding="utf-8") as file:
        for line in file:
            words = line.split()

            for word in words:
                word = word.lower().strip(".,!?;:'\"()[]{}")
                word_count[word] = word_count.get(word, 0) + 1

    return word_count

word_frequency = count_word_frequency("sample.txt")
print(word_frequency)
````

## Explanation
- open the file in read mode
- read line by line
- split each line into words
- normalize words using lowercase and stripping punctuation
- count frequency using a dictionary

This is a very common interview pattern.


## Why `dict.get()` is useful here
Instead of writing:
```python
if word in word_count:
    word_count[word] += 1
else:
    word_count[word] = 1
```

we can write:
```python
word_count[word] = word_count.get(word, 0) + 1
```


# 8. Example 7: Validate an Email Address

This example uses:
- functions
- regular expressions

A basic valid email usually looks like:
```python
username@example.com
```


````python
import re

def is_valid_email(email):
    """Validate email address using a regular expression."""
    pattern = r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"
    return re.fullmatch(pattern, email) is not None

print(is_valid_email("elonmusk4688@gmail.com"))
print(is_valid_email("invalid-email"))
print(is_valid_email("user@site"))
````

## Explanation
The regex checks:
- username part before `@`
- domain name after `@`
- a dot `.` followed by at least 2 letters

### Important note
This is a **basic validation**, not a perfect universal validator for every valid email format in the real world.
