# 1. Introduction to Modules and Packages

Python has a huge ecosystem of reusable code.

Instead of writing everything from scratch, we can use:
- built-in modules
- third-party packages
- our own custom modules and packages

## Why modules and packages matter
They help with:
- organizing code
- reusing code
- reducing duplication
- improving maintainability
- structuring larger projects


# 2. What Is a Module?

A **module** is usually a single Python file containing reusable code.

Examples:
- functions
- variables
- classes

## Example
If you create a file named:
```python
math_utils.py
```
then that file itself is a module.


# 3. What Is a Package?

A **package** is a directory that groups multiple related modules together.

For example:
```python
my_package/
    __init__.py
    math_utils.py
    string_utils.py
```

This helps organize code into folders.


# 4. Importing Built-in Modules

Python comes with many built-in modules.

One common example is the `math` module.


````python
import math

print(math.sqrt(16))
print(math.sqrt(25))
print(math.pi)
````

## Explanation
- `import math` loads the `math` module
- `math.sqrt(16)` calls the square root function from that module
- `math.pi` accesses the constant `pi`


## Exploring what a module provides
You can inspect available names inside a module with `dir()`.


````python
import math

print(dir(math)[:20])  # showing only part of the result
````

# 5. Importing Specific Names from a Module

Instead of importing the full module, you can import only the things you need.


````python
from math import sqrt, pi

print(sqrt(16))
print(sqrt(25))
print(pi)
````

## Explanation
Now you can use `sqrt` and `pi` directly, without writing `math.` before them.


# 6. Import Aliases with `as`

You can use `as` to create a shorter name for a module.


````python
import math as m

print(m.sqrt(36))
print(m.pi)
````

## Why aliases are useful
Aliases are helpful when:
- module names are long
- a short standard alias is widely used
- code becomes cleaner


## Example: NumPy alias
A very common real-world example is:
```python
import numpy as np
```


# 7. Installing External Packages

Some modules are built in, but others must be installed manually.

For example, `numpy` is a third-party package.


## Typical installation command
```bash
pip install numpy
```

Or from a requirements file:
```bash
pip install -r requirements.txt
```


## Example import
This notebook avoids requiring NumPy at runtime, but this is the usual syntax:
```python

```


````python
# Example only:
example_import = "import numpy as np"
print(example_import)
````

# 8. Importing Everything with `*`

You can write:
```python
from math import *
```

This imports many names into the current namespace.


````python
from math import *

print(sqrt(49))
print(pi)
````

## Why `import *` is discouraged
Although it works, it is usually discouraged because:
- it can overwrite existing names
- it makes code harder to understand
- you cannot easily see where a name came from

### Prefer:
- `import math`
or
- `from math import sqrt, pi`

# 10. Creating Your Own Package

A package is a folder containing modules.

## Example structure
```python
package/
    __init__.py
    maths.py
```

### Example `maths.py`
```python
def addition(a, b):
    return a + b

def subtraction(a, b):
    return a - b
```


## Why `__init__.py` matters
Traditionally, `__init__.py` marks the folder as a package and can also run initialization code.

For beginner-friendly package structure, it is good practice to include it.


# 11. Importing from Your Own Package

If you have:
```python
package/
    __init__.py
    maths.py
```

Then you can import like this:
```python
from package.maths import addition
```


````python
# Demonstration as plain code example
example_import = "from package.maths import addition"
print(example_import)
````

## Or import the module itself
```python
from package import maths
print(maths.addition(2, 3))
```


````python
example_import_2 = "from package import maths"
print(example_import_2)
print("Then call: maths.addition(2, 3)")
````

# 12. Subpackages

A package can contain another package.

## Example structure
```python
package/
    __init__.py
    maths.py
    subpackage/
        __init__.py
        mult.py
```

### Example `mult.py`
```python
def multiply(a, b):
    return a * b
```


## Importing from a subpackage
```python
from package.subpackage.mult import multiply
```


````python
example_import_3 = "from package.subpackage.mult import multiply"
print(example_import_3)
print("Then call: multiply(4, 5)")
````

# 13. Example Project Structure

Here is a cleaner conceptual layout:

```python
project/
    main.py
    package/
        __init__.py
        maths.py
        subpackage/
            __init__.py
            mult.py
```

## Example contents

### `maths.py`
```python
def addition(a, b):
    return a + b

def subtraction(a, b):
    return a - b
```

### `mult.py`
```python
def multiply(a, b):
    return a * b
```

### `main.py`
```python
from package.maths import addition, subtraction
from package.subpackage.mult import multiply

print(addition(2, 3))
print(subtraction(4, 3))
print(multiply(4, 5))
```


## Package
A folder containing multiple related modules

Example:
```python
package/
    __init__.py
    maths.py

```


# 15. Common Import Styles

## Style 1: Import full module
```python
import math
math.sqrt(16)
```

## Style 2: Import specific names
```python
from math import sqrt, pi
sqrt(16)
```

## Style 3: Import with alias
```python
import math as m
m.sqrt(16)
```

## Style 4: Import everything
```python
from math import *
```
Usually discouraged.

# 17. Common Errors with Modules and Packages

## 1. `ModuleNotFoundError`
This happens when Python cannot find the module.

Example:
```python
import numpy
```
without installing NumPy may fail.

## 2. Wrong folder structure
If your files are not placed correctly, imports may fail.

## 3. Missing `__init__.py` in beginner-style package setups
This can cause confusion in package recognition, especially while learning structured imports.

## 4. Namespace confusion from `import *`
Too many imported names can make code hard to follow.

## 5. Importing from the wrong working directory
Sometimes the code is correct, but Python is running from a location where it cannot resolve the package path.


# 18. Best Practices

## 1. Prefer explicit imports
Good:
```python
from math import sqrt
```

## 2. Use aliases only when useful
Example:
```python
import numpy as np
```

## 3. Avoid `import *` in most real projects
It reduces clarity.

## 4. Organize large codebases into packages
Packages help keep related code together.

## 5. Keep related functions in related modules
For example:
- `maths.py` for math utilities
- `string_utils.py` for string logic

## 6. Use meaningful package and module names
Readable names improve maintainability.
