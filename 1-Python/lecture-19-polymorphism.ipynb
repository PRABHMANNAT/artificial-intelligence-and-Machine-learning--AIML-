# 1. Introduction to Polymorphism

Polymorphism is a core concept in object-oriented programming.

The word **polymorphism** means:
- **poly** = many
- **morphism** = forms

So polymorphism means:
> one interface, many forms

## Simple meaning
Different objects can respond to the **same method name** in different ways.


# 2. Why Polymorphism Matters

Polymorphism helps you write code that is:
- flexible
- reusable
- extensible
- easier to maintain

Instead of writing separate logic for every type, you can often call the same method name and let each object handle it in its own way.


# 3. Method Overriding

Method overriding happens when a child class provides its own implementation of a method that already exists in the parent class.

This is one of the most common ways polymorphism appears in OOP.


## Example: Base class `Animal`


````python
class Animal:
    def speak(self):
        return "Sound of the animal"


class Dog(Animal):
    def speak(self):
        return "Woof!"


class Cat(Animal):
    def speak(self):
        return "Meow!"


dog = Dog()
cat = Cat()

print(dog.speak())
print(cat.speak())
````

## Explanation
Here:
- `Animal` defines `speak()`
- `Dog` overrides `speak()`
- `Cat` overrides `speak()`

The method name is the same:
```python
speak()
```

But the behavior changes depending on the object.


# 4. Same Function, Different Object Types

A function can take different objects and call the same method on them.

This demonstrates polymorphism clearly.


````python
class Animal:
    def speak(self):
        return "Sound of the animal"


class Dog(Animal):
    def speak(self):
        return "Woof!"


class Cat(Animal):
    def speak(self):
        return "Meow!"


def animal_speak(animal):
    print(animal.speak())


dog = Dog()
cat = Cat()

animal_speak(dog)
animal_speak(cat)
````

## Explanation
The function:
```python
def animal_speak(animal):
    print(animal.speak())
```

does not care whether `animal` is:
- a `Dog`
- a `Cat`
- or some other object with a `speak()` method

It just calls `speak()`.


# 5. Polymorphism and Duck Typing

Python often uses **duck typing**.

The idea is:
> If it behaves like the required type, Python is happy to use it.

So in many cases, an object does not have to inherit from the same parent as long as it provides the expected method.


````python
class Dog:
    def speak(self):
        return "Woof!"


class Cat:
    def speak(self):
        return "Meow!"


class Bird:
    def speak(self):
        return "Chirp!"


def animal_speak(animal):
    print(animal.speak())


animal_speak(Dog())
animal_speak(Cat())
animal_speak(Bird())
````

## Explanation
Here polymorphism works even without a shared parent class, because each object supports the same method:
```python
speak()
```


# 6. Polymorphism with Functions and Methods

Polymorphism can also be demonstrated by designing classes that share the same method name and then processing them through a common function.


## Example: Shapes with `area()`


````python
import math

class Shape:
    def area(self):
        return "Area of the figure"


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius


def print_area(shape):
    print(f"Area is: {shape.area()}")


rectangle = Rectangle(4, 5)
circle = Circle(3)

print_area(rectangle)
print_area(circle)
````

## Explanation
- `Rectangle.area()` calculates area using width × height
- `Circle.area()` calculates area using πr²

The function:
```python
print_area(shape)
```
can work with both objects because both provide an `area()` method.


# 7. Parent Class with Placeholder Methods

Sometimes a base class includes a method that is meant to be overridden by child classes.


````python
class Shape:
    def area(self):
        raise NotImplementedError("Subclasses must implement the area() method.")


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height


rect = Rectangle(10, 2)
print(rect.area())
````

## Why this is useful
This clearly communicates:
> subclasses are expected to provide their own implementation


# 8. Polymorphism with Abstract Base Classes

Abstract base classes (ABCs) are a more formal way to define common interfaces.

They help enforce that derived classes implement certain methods.


## Example: Vehicle abstract base class


````python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass


class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"


class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle engine started"


def start_vehicle(vehicle):
    print(vehicle.start_engine())


car = Car()
motorcycle = Motorcycle()

start_vehicle(car)
start_vehicle(motorcycle)
````

## Explanation
Here:
- `Vehicle` is an abstract base class
- `start_engine()` is an abstract method
- child classes **must** implement `start_engine()`

This enforces consistency across implementations.


# 9. Why Abstract Base Classes Matter

ABCs are useful when you want:
- a clear contract
- required methods
- consistent structure across subclasses

They are especially useful in larger codebases.


## What happens if a child class does not implement the abstract method?
Python prevents creating its object.


````python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass


class IncompleteVehicle(Vehicle):
    pass

# Uncomment to see the error:
# obj = IncompleteVehicle()
````

# 10. Comparing the Three Polymorphism Styles Shown Here

## 1. Method overriding
Child classes redefine the parent method.

## 2. Duck typing with common method names
Different classes support the same method without necessarily sharing a parent.

## 3. Abstract base classes
A formal base class enforces that subclasses implement required methods.


# 11. Practical Example: Animal Sounds

This is the simplest and clearest polymorphism example.


````python
class Animal:
    def speak(self):
        return "Animal sound"


class Dog(Animal):
    def speak(self):
        return "Woof!"


class Cat(Animal):
    def speak(self):
        return "Meow!"


class Cow(Animal):
    def speak(self):
        return "Moo!"


animals = [Dog(), Cat(), Cow()]

for animal in animals:
    print(animal.speak())
````

## Key idea
Same method:
```python
speak()
```

Different output depending on object type.


# 12. Practical Example: Shapes and Area

This is another classic polymorphism example.


````python
import math

class Shape:
    def area(self):
        raise NotImplementedError("Subclasses must implement area()")


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


shapes = [Rectangle(4, 5), Circle(3), Triangle(10, 2)]

for shape in shapes:
    print(shape.area())
````

# 13. Practical Example: Vehicles and Engines

This uses abstract base classes.


````python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass


class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"


class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle engine started"


class ElectricScooter(Vehicle):
    def start_engine(self):
        return "Electric scooter powered on"


vehicles = [Car(), Motorcycle(), ElectricScooter()]

for vehicle in vehicles:
    print(vehicle.start_engine())
````

# 14. Common Errors in Polymorphism

## 1. Forgetting to override a method
If the base method is only a placeholder, you may get:
- wrong generic output
- `NotImplementedError`
- ABC instantiation errors

## 2. Calling a method that an object does not have
If your function expects `obj.area()` but the object has no `area()`, it fails.

## 3. Using abstract classes incorrectly
If a subclass does not implement required abstract methods, Python blocks object creation.

## 4. Confusing overloading with overriding
Python polymorphism in these examples is mainly about overriding, not classic compile-time overloading.

## 5. Returning inconsistent kinds of results
If one class returns a number and another returns a string for the same conceptual method, code may become harder to use.


# 15. Best Practices

## 1. Keep a common interface
If objects are meant to be used polymorphically, method names should be consistent.

## 2. Use method overriding clearly
Child classes should provide meaningful specialized behavior.

## 3. Use abstract base classes when you need strict contracts
They make interfaces more explicit.

## 4. Prefer readability
Polymorphism is powerful, but your design should stay easy to understand.

## 5. Let shared functions depend on behavior, not exact class names
This is one of the biggest strengths of polymorphism.
