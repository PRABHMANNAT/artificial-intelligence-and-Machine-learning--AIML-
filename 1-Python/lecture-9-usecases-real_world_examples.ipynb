# 2. Real-World Example 1: Manage a To-Do List

A to-do app often needs to:
- store pending tasks
- add new tasks
- remove completed tasks
- check if a task still exists
- display remaining tasks


````python
# Initial tasks
to_do_list = ["Buy groceries", "Clean the house", "Pay bills"]

# Add new tasks
to_do_list.append("Schedule meeting")
to_do_list.append("Go for a run")

# Mark one task as completed
to_do_list.remove("Clean the house")

# Check if a task still exists
if "Pay bills" in to_do_list:
    print("Reminder: Don't forget to pay the utility bills.")

# Display remaining tasks
print("Remaining to-do list:")
for task in to_do_list:
    print(f"- {task}")
````

## What this example teaches
This example shows common list operations:
- `append()` to add tasks
- `remove()` to delete completed tasks
- `in` for membership checking
- `for` loop to display tasks


## Why lists fit this use case
A to-do list changes over time:
- tasks are added
- tasks are removed
- tasks are checked

Because lists are mutable, they are a natural fit for this kind of application state.


## Cleaner reusable version


````python
def add_task(tasks, task):
    tasks.append(task)

def complete_task(tasks, task):
    if task in tasks:
        tasks.remove(task)

def show_tasks(tasks):
    print("Tasks remaining:")
    for t in tasks:
        print(f"- {t}")

tasks = ["Buy groceries", "Clean the house", "Pay bills"]

add_task(tasks, "Schedule meeting")
add_task(tasks, "Go for a run")
complete_task(tasks, "Clean the house")
show_tasks(tasks)
````

# 3. Real-World Example 2: Organizing Student Grades

A list can also be used to store student grades.

Typical tasks:
- add a new grade
- compute average
- find highest grade
- find lowest grade


````python
grades = [85, 92, 78, 90, 88]

# Add a new grade
grades.append(95)

# Calculate average
average_grade = sum(grades) / len(grades)

# Find highest and lowest grade
highest_grade = max(grades)
lowest_grade = min(grades)

print("Grades:", grades)
print("Average grade:", average_grade)
print("Highest grade:", highest_grade)
print("Lowest grade:", lowest_grade)
````

## What this example teaches
This example combines:
- list storage
- `append()`
- `sum()`
- `len()`
- `max()`
- `min()`


## Why lists fit this use case
Grades are naturally a collection of values.  
A list allows easy:
- insertion of new grades
- summary calculation
- iteration and reporting


## Slightly improved reporting example


````python
grades = [85, 92, 78, 90, 88, 95]

print("Student grade report")
print("--------------------")
print("Grades:", grades)
print(f"Total subjects: {len(grades)}")
print(f"Average: {sum(grades) / len(grades):.2f}")
print(f"Highest: {max(grades)}")
print(f"Lowest: {min(grades)}")
````

# 4. Real-World Example 3: Managing Inventory

An inventory list can track which items are currently available.

Typical operations:
- add a new item to stock
- remove an item when sold
- check whether an item is available
- display updated inventory


````python
inventory = ["Apples", "Bananas", "Oranges", "Grapes"]

# New stock arrives
inventory.append("Strawberries")

# Item sold
inventory.remove("Bananas")

# Check stock availability
item = "Oranges"

if item in inventory:
    print(f"{item} are in stock.")
else:
    print(f"{item} are out of stock.")

# Display current inventory
print("Current inventory:")
for product in inventory:
    print(f"- {product}")
````

## What this example teaches
This example again shows that lists are useful when items are:
- added
- removed
- searched
- displayed


## Why lists fit this use case
Lists are useful for simple inventory prototypes or small apps.  
In larger systems, inventory would usually be stored in databases, but lists are still a great starting point for logic building.


## Safer inventory removal version
Using `remove()` on a missing item causes an error, so checking first is a good habit.


````python
inventory = ["Apples", "Bananas", "Oranges", "Grapes"]

sold_item = "Bananas"

if sold_item in inventory:
    inventory.remove(sold_item)

print(inventory)
````

# 5. Real-World Example 4: Collecting User Feedback

A list can also be used to store comments or feedback messages from users.

Possible tasks:
- store all feedback messages
- add new feedback
- count positive comments
- print all feedback


````python
feedbacks = [
    "Great service",
    "Very satisfied",
    "Could be better",
    "Excellent experience"
]

# Add new feedback
feedbacks.append("Not happy with the service")

# Count positive feedback
positive_feedback_count = sum(
    1
    for comment in feedbacks
    if "great" in comment.lower() or "excellent" in comment.lower()
)

print("Positive feedback count:", positive_feedback_count)

print("User feedback:")
for comment in feedbacks:
    print(f"- {comment}")
````

## What this example teaches
This example introduces:
- storing multiple text comments in a list
- appending new user input
- using a loop to count based on conditions
- basic text analysis with `.lower()`


## Why lists fit this use case
Feedback naturally arrives as multiple comments.  
A list is a simple and effective way to collect and process them.


## Improved sentiment keyword check
You can expand the positive keyword list to make the idea more realistic.


````python
feedbacks = [
    "Great service",
    "Very satisfied",
    "Could be better",
    "Excellent experience",
    "Not happy with the service",
    "Amazing support"
]

positive_words = ["great", "excellent", "amazing", "satisfied"]

positive_feedback_count = 0

for comment in feedbacks:
    lowered = comment.lower()
    if any(word in lowered for word in positive_words):
        positive_feedback_count += 1

print("Positive feedback count:", positive_feedback_count)
````

# 8. Common Mistakes in Real-World List Usage

## 1. Removing an item that does not exist
Using:
```python
my_list.remove("x")
```
on a missing value raises `ValueError`.

## 2. Assuming a task/item exists without checking
Always check with `in` if needed.

## 3. Dividing by zero when calculating averages
If a list is empty:
```python
sum(values) / len(values)
```
will fail.

## 4. Mutating lists without understanding the effect
Appending or removing changes the original list.

## 5. Confusing lists with sets
Lists allow duplicates and preserve order.  
Sets remove duplicates and are unordered.


````python
# Example of safe average calculation
grades = []

if grades:
    avg = sum(grades) / len(grades)
else:
    avg = 0

print("Average:", avg)
````

# 9. Best Practices

## 1. Use meaningful variable names
Good:
```python
to_do_list
inventory
grades
feedbacks
```

## 2. Check before removing
Prefer:
```python
if item in inventory:
    inventory.remove(item)
```

## 3. Handle empty lists safely
Especially when calculating averages or extremes.

## 4. Use functions to organize repeated logic
This makes code easier to test and reuse.

## 5. Keep list logic simple
Lists are great for small to medium in-memory collections.

## 6. Move to dictionaries or databases when structure grows
Real applications often combine lists with more advanced data structures.

# 12. Tricky Questions

### 1. What happens if you calculate average on an empty grades list?
**Answer:**  
It causes an error due to division by zero, unless handled safely.

### 2. Does `append()` return the modified list?
**Answer:**  
No. It modifies the list in-place and returns `None`.

### 3. What is the difference between `in` on a list and on a set?
**Answer:**  
Both check membership, but sets are generally faster for membership testing.

### 4. What happens if feedback matching is case-sensitive?
**Answer:**  
You may miss valid matches like `"Great"` vs `"great"`, which is why `.lower()` is useful.

### 5. Can a list be used for inventory in a huge production system by itself?
**Answer:**  
Usually not by itself. Large systems typically use databases and richer structures, but list logic is still foundational.

### 6. What is the difference between `remove()` and `pop()`?
**Answer:**  
`remove()` deletes by value, while `pop()` removes by index and returns the removed item.

### 7. Why is `max()` useful in grade analysis?
**Answer:**  
It quickly finds the highest score.

### 8. Why is `len()` useful in list-based apps?
**Answer:**  
It helps count remaining tasks, number of grades, number of comments, and more.
