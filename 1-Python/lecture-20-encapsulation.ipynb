## 2) What Is Encapsulation?

Encapsulation is the idea of:

- wrapping **data** (variables / attributes)
- and the **methods** that operate on that data

inside a single unit, usually a **class**.

It also involves **controlling access** to internal object data so that it is not changed carelessly from outside.

### Simple meaning
Instead of letting everything be directly modified from anywhere, we decide:
- what should be public
- what should be internal
- what should be accessed only through methods


## 3) Real-World Analogy

Think of a **washing machine**.

From outside, you can see:
- buttons
- settings
- where to put clothes

But the internal motor logic, water flow mechanism, and timing system are hidden.

You interact through a controlled interface.

That is similar to encapsulation:
- internal details are protected
- only necessary interaction is exposed


## 4) Public, Protected, and Private in Python

The transcript introduces three kinds of access levels:

- **public**
- **protected**
- **private**

### Important Python clarification
In languages like Java/C++/C#, access modifiers are stricter.

In Python:
- these are mostly **naming conventions**
- not hard access restrictions in the same strict sense

Still, they are extremely important for code readability and design.


## 5) Public Attributes

A **public** attribute is a normal instance variable.

### Example
```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

Here:
- `self.name`
- `self.age`

are public attributes.

That means they can be accessed directly from outside the class.


````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person = Person("Krish", 34)

print(person.name)
print(person.age)
````

## 6) Why Public Attributes Are Called Public

These values are directly accessible from outside the class.

Example:
```python
person = Person("Krish", 34)
print(person.name)
```

This works because `name` is public.

### Meaning
Public attributes are easy to access and easy to modify.

This is convenient, but it may be risky if the attribute should not be freely changed.

## 8) Why We Sometimes Avoid Direct Access

Sometimes an attribute is important and should not be modified directly by any code from outside.

Examples:
- age should not become negative
- account balance should not be changed carelessly
- password should not be directly visible
- internal status should only be updated under conditions

That is where encapsulation becomes useful.


## 9) Private Attributes with Double Underscore

In Python, a common way to indicate a **private** attribute is to use **double underscore**:

```python
self.__name = name
self.__age = age
```

### Example
```python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
```

Now these are treated as private-style attributes.


````python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

person = Person("Krish", 34)
print(dir(person))
````

## 10) Important Clarification: Python Private Is Name Mangling

Double underscore does **not** make attributes absolutely inaccessible.

Instead, Python performs **name mangling**.

For example:

```python
self.__age
```

inside class `Person` becomes something like:

```python
_Person__age
```

So Python is not creating a perfectly hidden secret variable.  
It is making accidental direct access harder and signaling:

> "This is internal to the class. Do not touch casually."


## 11) Why `person.__name` Usually Fails

Suppose you write:

```python
print(person.__name)
```

This usually raises an error like:

```python
AttributeError
```

because the actual stored name is mangled internally, such as:

```python
_Person__name
```

### Key idea
Private in Python means:
- avoid direct outside usage
- use class methods to access safely


````python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

person = Person("Krish", 34)

# print(person.__name)  # This would usually raise AttributeError
print(person._Person__name)  # Technically possible, but bad practice
````

## 12) Protected Attributes with Single Underscore

A **protected-style** attribute is written with a **single underscore**:

```python
self._name = name
```

### Meaning in Python
This is mainly a convention that says:

> "This attribute is intended for internal use or subclasses. Please do not use it directly from outside unless necessary."

### Important clarification
Unlike stricter languages, Python still allows access from outside:

```python
print(obj._name)
```

This works technically, but it is considered less appropriate by convention.


````python
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

person = Person("Krish", 34)

print(person._name)  # Possible, but convention says internal/protected-style use
````

## 13) Public vs Protected vs Private Summary

| Style | Syntax | Meaning in Python |
|---|---|---|
| Public | `name` | Accessible from anywhere |
| Protected-style | `_name` | Internal-use convention; subclasses may use it |
| Private-style | `__name` | Name-mangled to discourage direct outside access |

### Important exam/interview note
Python does not enforce access control as strictly as Java/C++.
It relies heavily on **convention**.


## 14) Getter and Setter Methods

The transcript explains encapsulation using **getter** and **setter** methods.

### Why use them?
Because private data should not be directly accessed or modified from outside.

Instead, we expose methods such as:
- `get_name()`
- `set_name()`
- `get_age()`
- `set_age()`

This allows:
- controlled access
- validation
- safer updates


## 15) Example: Encapsulation with Getter and Setter

```python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative")
```

This is classic encapsulation style.


````python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative")

person = Person("Krish", 34)

print(person.get_name())
print(person.get_age())

person.set_age(35)
print(person.get_age())

person.set_age(-5)
````

## 16) Why Getter and Setter Help

With direct public access:

```python
person.age = -100
```

Nothing prevents bad data unless you write extra logic.

With a setter:

```python
person.set_age(-100)
```

you can validate it and reject invalid values.

### Benefit
Encapsulation allows the class to protect its own data.


## 17) Accessing Private Data Inside the Class

Private variables are meant to be used **inside the class** through methods.

Example:
```python
def get_name(self):
    return self.__name
```

This works because the class itself can access its private attributes.

### Important point
Private does not mean the class itself cannot use it.  
It mainly means external code should not access it directly.


## 18) Encapsulation and Inheritance

The transcript also mentions protected variables being accessible in derived classes.

### Cleaner Python explanation
- `_name` → protected-style by convention; intended for subclass/internal use
- `__name` → private-style; name-mangled, so subclasses do not access it directly in the same way

### Example with protected-style attribute
```python
class Person:
    def __init__(self, name):
        self._name = name

class Employee(Person):
    def show_name(self):
        return self._name
```

This is a typical use of protected-style attributes.


````python
class Person:
    def __init__(self, name, age, gender):
        self._name = name
        self._age = age
        self.gender = gender

class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)

    def show_name(self):
        return self._name

employee = Employee("Krish", 34, "Male")
print(employee.show_name())
````

## 19) Important Correction About Protected in Python

The transcript says protected variables cannot be accessed outside the class.

In strict OOP languages, that is closer to true.

In Python, however:

```python
print(employee._name)
```

is still possible.

### So what does `_name` really mean?
It means:

> "This is intended for internal/subclass use. Please avoid direct outside use."

So it is a **convention**, not a strict restriction.


## 20) Important Correction About Private in Python

The transcript says private variables cannot be accessed outside the class.

Practically, direct access like `person.__name` fails, which supports the idea.

However, because of name mangling, technically one can still access:

```python
person._Person__name
```

### Best interpretation
Private attributes in Python are:
- not meant for direct outside use
- protected through name mangling
- designed to prevent accidental misuse, not to create absolute security


## 21) A Better Modern Pythonic Approach: `@property`

The transcript teaches getter and setter methods, which are good for learning encapsulation.

In modern Python, a more Pythonic pattern is often to use `@property`.

### Example
```python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, value):
        if value > 0:
            self.__age = value
        else:
            raise ValueError("Age must be positive")
```

This allows clean syntax like:
```python
person.age
person.age = 35
```

while still controlling access.


````python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, value):
        if value > 0:
            self.__age = value
        else:
            raise ValueError("Age must be positive")

person = Person("Krish", 34)
print(person.age)

person.age = 35
print(person.age)
````

## 22) Why Encapsulation Improves Maintainability

Encapsulation makes code easier to maintain because:
- internal details stay controlled
- object state cannot be changed carelessly
- validation rules stay inside the class
- outside code depends on methods/interface, not internals

This helps when projects become larger.


## 23) Common Mistakes to Avoid

1. **Assuming Python access levels are strict like Java/C++**
   - Python mainly uses conventions

2. **Using double underscore everywhere unnecessarily**
   - Not every variable needs to be private-style

3. **Accessing mangled private names from outside**
   - technically possible, but poor practice

4. **Not validating values in setter methods**
   - then encapsulation loses part of its value

5. **Using public attributes for highly sensitive internal state**
   - risky if direct modification should be avoided

6. **Confusing abstraction and encapsulation**
   - they are related but different concepts


## 24) Encapsulation vs Abstraction

Even though this notebook focuses on encapsulation, it helps to compare them clearly.

### Encapsulation
Focuses on:
- bundling data and methods together
- controlling access to internal state

### Abstraction
Focuses on:
- hiding complexity
- exposing only necessary functionality

### Easy difference
- Encapsulation = **data protection / controlled access**
- Abstraction = **hiding implementation complexity**

## 28) Tricky / Interview Concepts

### Q1. Are access modifiers truly enforced in Python?
**Answer:**  
Not in the same strict way as Java/C++/C#. Python mainly uses naming conventions and name mangling.

---

### Q2. What is the difference between `_x` and `__x`?
**Answer:**  
- `_x` → protected-style convention
- `__x` → private-style name mangling

---

### Q3. Why does Python use conventions instead of strict enforcement?
**Answer:**  
Python follows a philosophy of developer responsibility and readability over heavy restriction.

---

### Q4. What is the biggest advantage of encapsulation?
**Answer:**  
Controlled access to internal state and easier maintenance of class invariants.

---

### Q5. What is a class invariant in simple terms?
**Answer:**  
A rule that should always remain true for valid object state, such as `age > 0`.

---

### Q6. Why should setter methods validate input?
**Answer:**  
Because they protect object state from becoming invalid.

---

### Q7. Is direct access always bad?
**Answer:**  
Not always. Public fields are fine when unrestricted access is acceptable. Encapsulation matters more for sensitive or rule-driven state.

---

### Q8. Why are getters/setters sometimes criticized in Python?
**Answer:**  
Because blindly writing Java-style getters/setters for everything is not always Pythonic. Often `@property` is cleaner.

---

### Q9. What does encapsulation protect against?
**Answer:**  
Mostly accidental misuse, careless updates, and weak design. It is not a security boundary.

---

### Q10. How is encapsulation related to abstraction?
**Answer:**  
Both hide internals in some way, but encapsulation is about controlled access to state, while abstraction is about hiding complexity.
