# 1. Introduction to Control Flow

## What is control flow?
Control flow means the order in which the program executes statements.

In Python, control flow helps decide:
- which block of code should run
- how many times code should run
- whether a condition is true or false

## Major parts of control flow
- conditional statements
- loops
- function calls
- exception handling

This notebook focuses on **conditional statements**.

# 3. The `if` Statement

The `if` statement checks a condition.

If the condition is `True`, Python executes the block inside the `if` statement.
If the condition is `False`, Python skips that block.


## Syntax
```python
if condition:
    # code block
```


````python
age = 20

if age >= 18:
    print("You are allowed to vote.")
````

## Explanation
In the above example:
- the condition is `age >= 18`
- since `age` is `20`, the condition is `True`
- therefore the `print()` statement runs


````python
age = 16

if age >= 18:
    print("You are allowed to vote.")

print("Program continues...")
````

### What happens here?
Since `age = 16`, the condition `age >= 18` is false.
So the indented block inside `if` does not run.
The program simply continues to the next line outside the `if`.


# 4. The `if-else` Statement

The `else` block runs when the `if` condition is false.

## Syntax
```python
if condition:
    # code if true
else:
    # code if false
```


````python
age = 16

if age >= 18:
    print("You are eligible for voting.")
else:
    print("You are a minor.")
````

## Explanation
- If age is 18 or more → first block runs
- otherwise → `else` block runs


### Key point
`else` does not check a separate condition.
It is the default fallback block.


# 5. The `if-elif-else` Statement

Use `elif` when you want to test multiple conditions in sequence.

## Syntax
```python
if condition1:
    # block 1
elif condition2:
    # block 2
elif condition3:
    # block 3
else:
    # default block
```


````python
age = 20

if age < 13:
    print("You are a child.")
elif age < 18:
    print("You are a teenager.")
else:
    print("You are an adult.")
````

## Explanation
Python checks conditions from top to bottom:

1. Is `age < 13`?
2. If not, is `age < 18`?
3. If not, run `else`

Only the **first matching block** executes.


````python
for age in [10, 15, 22]:
    print(f"Age = {age}")
    if age < 13:
        print("Child")
    elif age < 18:
        print("Teenager")
    else:
        print("Adult")
    print("---")
````

## Important behavior
In an `if-elif-else` chain:
- conditions are checked in order
- once one condition matches, the remaining conditions are skipped


# 6. Nested Conditional Statements

A nested conditional means placing one conditional statement inside another.

## Example shape
```python
if condition1:
    if condition2:
        ...
    else:
        ...
else:
    ...
```

This is called **nesting** because one `if` exists inside another block.


## When to use nested conditionals
Nested conditionals are useful when:
- a second decision depends on the first
- you want to break logic into levels

But if logic can be written more clearly with `elif`, prefer that.


````python
number = 12

if number > 0:
    print("The number is positive.")
    if number % 2 == 0:
        print("The number is even.")
    else:
        print("The number is odd.")
else:
    print("The number is zero or negative.")
````

## Explanation
Outer condition:
- checks whether the number is positive

Inner condition:
- only runs if the number is positive
- then checks whether it is even or odd

# 10. Practical Example 3: Leap Year Check

The leap-year logic is a classic interview question.

## Correct rule
A year is a leap year if:
- divisible by 400, or
- divisible by 4 but **not** by 100


````python
def is_leap_year(year: int) -> bool:
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

for y in [2022, 2024, 1900, 2000]:
    print(y, "->", is_leap_year(y))
````

## Explanation
- `2022` → not divisible by 4 → not leap year
- `2024` → divisible by 4, not by 100 → leap year
- `1900` → divisible by 100, not by 400 → not leap year
- `2000` → divisible by 400 → leap year

# 12. Practical Example 5: Ticket Price Based on Age and Student Status

This is a good example of combining multiple conditions.


````python
def ticket_price(age: int, is_student: bool) -> str:
    if age < 5:
        return "Free"
    elif age < 12:
        return "$10"
    elif age < 18:
        if is_student:
            return "$12"
        else:
            return "$15"
    else:
        if is_student:
            return "$18"
        else:
            return "$20"

examples = [
    (4, False),
    (10, False),
    (16, True),
    (16, False),
    (25, True),
    (25, False),
]

for age, student in examples:
    print(f"age={age}, student={student} -> {ticket_price(age, student)}")
````

# 13. Best Practices for Conditional Statements

## 1. Keep conditions readable
Prefer:
```python
if age >= 18:
```
over overly complex expressions packed into one line.

## 2. Use meaningful variable names
Good:
```python
is_student = True
```
Bad:
```python
x = True
```

## 3. Avoid too much nesting
If nesting becomes deep, consider:
- breaking logic into functions
- using `elif`
- simplifying conditions

## 4. Be careful with order
In an `if-elif-else` chain, order matters.
More specific conditions often should come before broader ones.

## 5. Always use proper indentation
Python relies on indentation to define blocks.

## 6. Handle edge cases
Examples:
- division by zero
- zero age / invalid age
- unexpected input


# 14. Common Errors and Mistakes

## 1. Missing colon `:`
Incorrect:
```python
if age >= 18
    print("Eligible")
```

Correct:
```python
if age >= 18:
    print("Eligible")
```


## 2. IndentationError
Incorrect indentation causes Python to fail.

Incorrect:
```python
if age >= 18:
print("Eligible")
```

Correct:
```python
if age >= 18:
    print("Eligible")
```


## 3. Using `=` instead of `==`
Incorrect:
```python
if age = 18:
```

Correct:
```python
if age == 18:
```


## 4. Wrong order of conditions
Example:
```python
age = 10

if age < 18:
    print("Minor")
elif age < 13:
    print("Child")
```

The second condition never runs for a child because `age < 18` already matched first.


## 5. Over-nesting
Too much nesting makes code hard to read and debug.

# 18. Tricky Questions

### 1. What is the difference between `if` and `elif`?
**Answer:**  
`if` starts the conditional chain. `elif` adds another condition after the previous one fails.

### 2. Will both blocks run here?
```python
x = 5
if x > 3:
    print("A")
elif x > 1:
    print("B")
```
**Answer:**  
No. Only `A` prints because once the first condition is true, the rest are skipped.

### 3. What is wrong with this logic?
```python
age = 10
if age < 18:
    print("Minor")
elif age < 13:
    print("Child")
```
**Answer:**  
The child case will never execute because all children under 13 are already covered by `age < 18`.

### 4. Does `else` need a condition?
**Answer:**  
No. `else` is the fallback block.

### 5. Can you put an `if` inside an `if`?
**Answer:**  
Yes. That is called nesting.

### 6. What is the result of:
```python
if 0:
    print("Yes")
else:
    print("No")
```
**Answer:**  
`No`  
Because `0` is treated as falsey.

### 7. What is the result of:
```python
if "hello":
    print("Yes")
```
**Answer:**  
`Yes`  
Because a non-empty string is truthy.

### 8. Can nested conditionals always be replaced with simpler logic?
**Answer:**  
Not always, but often code can be simplified using `elif`, helper variables, or functions.


# 19. Interview Concepts and Model Answers

### 1. What are conditional statements in Python?
