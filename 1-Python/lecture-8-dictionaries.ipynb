# 1. Introduction to Dictionaries

A **dictionary** is one of the most important Python data structures.

## Definition
A dictionary is an **unordered mapping of key-value pairs**.

### What this means
- each item has a **key**
- each key maps to a **value**
- keys are used to look up values

## Why dictionaries are useful
Dictionaries are useful when data has labels, such as:
- student information
- user profiles
- product details
- configuration settings
- database-like records


## Key properties of dictionaries

### 1. Key-value pairs
Every dictionary entry looks like:
```python
key: value
```

### 2. Keys must be unique
A dictionary cannot keep duplicate keys.

### 3. Keys must be hashable
Examples of valid keys:
- strings
- numbers
- tuples of hashable values

### 4. Values can be any type
A value can be:
- string
- number
- list
- tuple
- set
- dictionary
- any Python object


# 2. Creating Dictionaries

## Empty dictionary using `{}` 


````python
empty_dict = {}
print(empty_dict)
print(type(empty_dict))
````

## Empty dictionary using `dict()`

## Dictionary with key-value pairs


````python
student = {
    "name": "Krish",
    "age": 32,
    "grade": "A"
}

print(student)
print(type(student))
````

## Important note about duplicate keys
If the same key is repeated, the last value overrides the earlier one.


````python
student = {
    "name": "Krish",
    "age": 32,
    "name": "A"
}

print(student)
````

### Explanation
Here, `"name"` appears twice.  
The dictionary keeps only one `"name"` key, and its final value becomes `"A"`.


# 3. Accessing Dictionary Elements

There are two common ways to access dictionary values:

1. direct key access using square brackets
2. using the `get()` method


## 3.1 Direct key access
Syntax:
```python
student["grade"]
```


````python
student = {
    "name": "Krish",
    "age": 32,
    "grade": "A"
}

print(student["grade"])
print(student["age"])
````

### Important note
If the key does not exist, direct access raises `KeyError`.


````python
student = {
    "name": "Krish",
    "age": 32,
    "grade": "A"
}

# Uncomment to test:
# print(student["last_name"])
````

## 3.2 Using `get()`
Syntax:
```python
student.get("grade")
```

If the key is missing, `get()` returns `None` by default.


````python
student = {
    "name": "Krish",
    "age": 32,
    "grade": "A"
}

print(student.get("grade"))
print(student.get("last_name"))
````

## Using `get()` with a default value
You can provide a fallback value if the key is missing.


````python
student = {
    "name": "Krish",
    "age": 32,
    "grade": "A"
}

print(student.get("last_name", "Not Available"))
````

### Difference between direct access and `get()`
- `student["x"]` → raises error if key is missing
- `student.get("x")` → safer, returns `None` or your default value


# 4. Modifying Dictionary Elements

Dictionaries are mutable.

This means you can:
- update an existing value
- add a new key-value pair
- delete a key-value pair


## 4.1 Update an existing value


````python
student = {
    "name": "Krish",
    "age": 32,
    "grade": "A"
}

student["age"] = 33
print(student)
````

## 4.2 Add a new key-value pair


````python
student = {
    "name": "Krish",
    "age": 33,
    "grade": "A"
}

student["address"] = "India"
print(student)
````

## 4.3 Delete a key-value pair using `del`


````python
student = {
    "name": "Krish",
    "age": 33,
    "grade": "A",
    "address": "India"
}

del student["grade"]
print(student)
````

## 4.4 Remove and return a value using `pop()`


````python
student = {
    "name": "Krish",
    "age": 33,
    "grade": "A",
    "address": "India"
}

removed = student.pop("grade")
print("Removed value:", removed)
print("Updated dictionary:", student)
````

# 5. Common Dictionary Methods

Some of the most useful dictionary methods are:
- `keys()`
- `values()`
- `items()`
- `copy()`
- `update()`
- `pop()`
- `popitem()`
- `clear()`


## 5.1 `keys()`
Returns a view of all dictionary keys.


````python
student = {
    "name": "Krish",
    "age": 33,
    "address": "India"
}

keys = student.keys()
print(keys)
print(type(keys))
````

## 5.2 `values()`
Returns a view of all dictionary values.


````python
student = {
    "name": "Krish",
    "age": 33,
    "address": "India"
}

values = student.values()
print(values)
print(type(values))
````

## 5.3 `items()`
Returns a view of key-value pairs.


````python
student = {
    "name": "Krish",
    "age": 33,
    "address": "India"
}

items = student.items()
print(items)
print(type(items))
````

## 5.4 `update()`
Adds multiple key-value pairs from another dictionary.


````python
student = {
    "name": "Krish",
    "age": 33
}

student.update({"grade": "A", "address": "India"})
print(student)
````

Removes and returns the last inserted key-value pair.


````python
student = {
    "name": "Krish",
    "age": 33,
    "grade": "A"
}

removed_item = student.popitem()
print("Removed item:", removed_item)
print("Remaining:", student)
````

## 5.6 `clear()`
Removes all items from the dictionary.


````python
student = {
    "name": "Krish",
    "age": 33,
    "grade": "A"
}

student.clear()
print(student)
````

# 6. Shallow Copy and Reference Behavior

This is one of the most important concepts in dictionary handling.


## Direct assignment does NOT create a separate copy


````python
student = {
    "name": "Krish",
    "age": 33,
    "address": "India"
}

student_copy = student

print("Original:", student)
print("Assigned copy:", student_copy)

student["name"] = "Krish 2"

print("After modifying original:")
print("Original:", student)
print("Assigned copy:", student_copy)
````

### Explanation
Both variables point to the **same dictionary object**.
So changes through one reference are visible through the other.


## Using `copy()` to create a shallow copy


````python
student = {
    "name": "Krish 2",
    "age": 33,
    "address": "India"
}

student_copy_1 = student.copy()

print("Original:", student)
print("Shallow copy:", student_copy_1)

student["name"] = "Krish 3"

print("After modifying original:")
print("Original:", student)
print("Shallow copy:", student_copy_1)
````

### Important note
`copy()` creates a **shallow copy**.

This means:
- top-level dictionary is copied
- nested mutable objects inside it are still shared


````python
student = {
    "name": "Krish",
    "skills": ["Python", "SQL"]
}

student_copy = student.copy()

student["skills"].append("Pandas")

print("Original:", student)
print("Shallow copy:", student_copy)
````

### Why both changed
Because the inner list `skills` is still the same shared object in both dictionaries.


# 7. Iterating Over Dictionaries

You can loop through:
- keys
- values
- key-value pairs


## 7.1 Iterate over keys


````python
student = {
    "name": "Krish 3",
    "age": 33,
    "address": "India"
}

for key in student.keys():
    print(key)
````

## 7.2 Iterate over values


````python
student = {
    "name": "Krish 3",
    "age": 33,
    "address": "India"
}

for value in student.values():
    print(value)
````

## 7.3 Iterate over key-value pairs


````python
student = {
    "name": "Krish 3",
    "age": 33,
    "address": "India"
}

for key, value in student.items():
    print(f"{key}: {value}")
````

## Default iteration behavior

```python

    ...
```
Python iterates over the keys.


````python
student = {
    "name": "Krish 3",
    "age": 33,
    "address": "India"
}

for x in student:
    print(x)
````

# 8. Nested Dictionaries

A nested dictionary is a dictionary inside another dictionary.


````python
students = {
    "student1": {"name": "Krish", "age": 32},
    "student2": {"name": "Peter", "age": 35}
}

print(students)
````

## Access nested dictionary elements


````python
students = {
    "student1": {"name": "Krish", "age": 32},
    "student2": {"name": "Peter", "age": 35}
}

print(students["student2"]["name"])
print(students["student2"]["age"])
````

## Explanation
- `students["student2"]` gives the inner dictionary
- `["name"]` or `["age"]` then accesses values inside it


# 9. Iterating Over Nested Dictionaries


````python
students = {
    "student1": {"name": "Krish", "age": 32},
    "student2": {"name": "Peter", "age": 35}
}

for student_id, student_info in students.items():
    print(student_id)
    for key, value in student_info.items():
        print(f"  {key}: {value}")
````

## Explanation
Outer loop:
- iterates over top-level student IDs and student info

Inner loop:
- iterates over key-value pairs inside each student's dictionary


# 10. Dictionary Comprehension

Dictionary comprehension is a concise way to create dictionaries.

## General syntax
```python
{key_expr: value_expr for item in iterable}
```

## With condition
```python
{key_expr: value_expr for item in iterable if condition}
```


## Example: squares dictionary


````python
squares = {x: x**2 for x in range(5)}
print(squares)
````

## Explanation
This creates:
```python
{0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
```


## Example: squares of only even numbers


````python
even_squares = {x: x**2 for x in range(10) if x % 2 == 0}
print(even_squares)
````

## Another example: word lengths


````python
words = ["apple", "banana", "kiwi"]

word_lengths = {word: len(word) for word in words}
print(word_lengths)
````

# 11. Practical Example: Frequency Count of Elements in a List

This is a very useful real-world pattern.


````python
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1

print(frequency)
````

## Explanation
- first time a number appears → set count to 1
- next time it appears → increment the existing count


## Cleaner alternative using `dict.get()`


````python
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

frequency = {}

for number in numbers:
    frequency[number] = frequency.get(number, 0) + 1

print(frequency)
````

# 12. Practical Example: Merge Two Dictionaries

You can merge dictionaries in multiple ways.


## Using `**` unpacking


````python
dictionary1 = {"a": 1, "b": 2}
dictionary2 = {"b": 3, "c": 4}

merged_dictionary = {**dictionary1, **dictionary2}
print(merged_dictionary)
````

### Important note
If the same key exists in both dictionaries, the value from the later dictionary wins.

So here:
- `"b": 2` from `dictionary1`
is overwritten by
- `"b": 3` from `dictionary2`


## Using the merge operator `|` (Python 3.9+)


````python
dictionary1 = {"a": 1, "b": 2}
dictionary2 = {"b": 3, "c": 4}

merged_dictionary = dictionary1 | dictionary2
print(merged_dictionary)
````
