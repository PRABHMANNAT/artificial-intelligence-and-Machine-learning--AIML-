# 1. What Is the Python Standard Library?

The **standard library** is the collection of modules that are included with Python by default.

## Why it matters
The standard library helps you perform many tasks without installing extra packages, such as:
- math operations
- file handling
- random values
- date and time handling

- regular expressions

# 2. `array` Module

The `array` module provides memory-efficient arrays of basic typed values.

## Why use it
Unlike Python lists, arrays from the `array` module store items of a **single type**.


````python
import array

arr = array.array('i', [1, 2, 3, 4])
print(arr)
print(type(arr))
````

## Explanation
- `'i'` is the type code for integers
- the second argument is an iterable of values
- the result is an array restricted to that type


## Common type codes
- `'i'` → integer
- `'f'` → float
- `'d'` → double-precision float


# 3. `math` Module

The `math` module provides mathematical functions and constants.


````python
import math

print("Square root of 16:", math.sqrt(16))
print("Square root of 25:", math.sqrt(25))
print("Pi value:", math.pi)
````

## Importing specific names
Instead of importing the full module, you can import only what you need.


````python
from math import sqrt, pi

print(sqrt(36))
print(pi)
````

## Commonly used `math` functions
- `math.sqrt(x)` → square root
- `math.ceil(x)` → round up
- `math.floor(x)` → round down
- `math.factorial(x)` → factorial
- `math.pi` → pi constant


````python
import math

print("ceil(4.2):", math.ceil(4.2))
print("floor(4.8):", math.floor(4.8))
print("factorial(5):", math.factorial(5))
````

# 4. `random` Module

The `random` module is used to generate random values.


````python
import random

print("Random integer between 1 and 10:", random.randint(1, 10))
print("Random choice:", random.choice(["apple", "banana", "cherry"]))
````

## Explanation
- `randint(a, b)` returns a random integer between `a` and `b`, inclusive
- `choice(seq)` picks one random element from a sequence


## More useful random functions
- `random.random()` → float between 0.0 and 1.0
- `random.shuffle(list)` → shuffles a list in place
- `random.sample(seq, k)` → picks `k` unique items


````python
import random

values = [1, 2, 3, 4, 5]
random.shuffle(values)

print("Shuffled:", values)
print("Sample of 2:", random.sample(values, 2))
print("Random float:", random.random())
````

# 5. `os` Module

The `os` module helps interact with the operating system.

It is commonly used for:
- current working directory
- creating folders
- checking paths
- file system operations


````python
import os

print("Current working directory:")
print(os.getcwd())
````

## Creating a directory


````python
import os
from pathlib import Path

demo_dir = Path("test_dir_standard_lib_demo")

if not demo_dir.exists():
    os.mkdir(demo_dir)

print("Directory exists:", demo_dir.exists())
````

## Useful `os` / `os.path` operations
- `os.getcwd()` → get current working directory
- `os.mkdir(name)` → create one directory
- `os.listdir(path)` → list files/folders
- `os.path.exists(path)` → check existence


````python
import os

print("Files/folders in current directory:")
print(os.listdir(".")[:10])
````

# 6. `shutil` Module

The `shutil` module helps with higher-level file operations, such as copying files.


````python
import shutil
from pathlib import Path

src = Path("source_demo.txt")
dst = Path("destination_demo.txt")

src.write_text("Hello world\nHow are you?\nWelcome to the course.", encoding="utf-8")
shutil.copyfile(src, dst)

print("Source exists:", src.exists())
print("Destination exists:", dst.exists())
print("Destination content:")
print(dst.read_text(encoding="utf-8"))
````

## Explanation
`shutil.copyfile(src, dst)` copies file contents from source to destination.


## Other useful `shutil` operations
- `shutil.copy()` → copy file with metadata behavior differences
- `shutil.move()` → move/rename files
- `shutil.rmtree()` → remove a directory tree


# 7. `json` Module

The `json` module is used for serialization and deserialization of JSON data.

## Why JSON matters
JSON is widely used in:
- APIs
- web applications
- configuration files
- NoSQL-style data exchange


````python
import json

data = {
    "name": "Krish",
    "age": 25
}

json_str = json.dumps(data)

print("JSON string:")
print(json_str)
print(type(json_str))
````

## Convert JSON string back to Python object


````python
import json

json_str = '{"name": "Krish", "age": 25}'
parsed_data = json.loads(json_str)

print(parsed_data)
print(type(parsed_data))
````

## Explanation
- `json.dumps()` → Python object to JSON string
- `json.loads()` → JSON string to Python object


## File-based JSON methods
- `json.dump(obj, file)` → write JSON to a file
- `json.load(file)` → read JSON from a file


````python
import json
from pathlib import Path

json_file = Path("sample_data.json")

data = {"course": "Python", "level": "beginner"}
with open(json_file, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2)

with open(json_file, "r", encoding="utf-8") as f:
    loaded = json.load(f)

print(loaded)
````

# 8. `csv` Module

The `csv` module is used to read and write CSV files.


````python
import csv

with open("example.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["name", "age"])
    writer.writerow(["Krish", 32])

with open("example.csv", "r", encoding="utf-8") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
````

## Explanation
- `csv.writer(file)` creates a CSV writer
- `writer.writerow(...)` writes one row
- `csv.reader(file)` reads rows back


## Why `newline=""` is used
When writing CSV files, `newline=""` helps avoid blank-line issues on some systems.


# 9. `datetime` Module

The `datetime` module helps work with dates and times.


````python
from datetime import datetime, timedelta

now = datetime.now()
print("Current date and time:", now)

yesterday = now - timedelta(days=1)
print("Yesterday:", yesterday)
````

## Explanation
- `datetime.now()` gives current local date and time
- `timedelta(days=1)` represents a time difference of one day
- subtracting the timedelta gives yesterday


## More useful datetime operations
- `datetime.today()`
- `strftime()` for formatting
- `strptime()` for parsing strings into dates


````python
from datetime import datetime

now = datetime.now()
formatted = now.strftime("%Y-%m-%d %H:%M:%S")

print("Formatted time:", formatted)
````

# 10. `time` Module

The `time` module provides time-related functions, including sleeping and raw timestamps.


````python
import time

start = time.time()
time.sleep(2)
end = time.time()

print("Start timestamp:", start)
print("End timestamp:", end)
print("Elapsed seconds:", end - start)
````

## Explanation
- `time.time()` returns seconds since the Unix epoch
- `time.sleep(2)` pauses the program for about 2 seconds


# 11. `re` Module (Regular Expressions)

The `re` module is used for pattern matching in strings.

It is very useful for:
- validation
- searching text
- extracting patterns
- cleaning data


````python
import re

pattern = r"\d+"
text = "There are 123 apples"

match = re.search(pattern, text)

print("Matched text:", match.group())
````

## Explanation
- `\d+` means one or more digits
- `re.search()` finds the first match in the text
- `match.group()` returns the matched substring


## Another example: find all digit groups


````python
import re

pattern = r"\d+"
text = "There are 123 apples and 456 oranges"

matches = re.findall(pattern, text)
print(matches)
````

## Common regex functions
- `re.search()` → first match anywhere
- `re.match()` → match only at the start
- `re.findall()` → all matches
- `re.sub()` → replace matches


````python
import re

text = "My phone number is 9876543210"
masked = re.sub(r"\d", "*", text)

print(masked)
````

# 14. Best Practices

## 1. Use standard library first when it fits
It is stable, well-tested, and already available.

## 2. Use specific imports when helpful
Example:
```python

```

## 3. Use module prefixes when clarity matters
Example:
```python

math.sqrt(16)
```

## 4. Prefer raw strings for regex
Example:
```python

```

## 5. Use context managers for files
Example:
```python

    ...
```

## 6. Be clear about data conversion
Know when you are working with:
- Python dict/list
- JSON string
- CSV rows
- timestamps
- datetime objects

# 17. Tricky Questions

### 1. Is `datetime` the same as `time`?
**Answer:**  
No. They are related but different modules and tools.

### 2. Does `time.time()` show a human-readable time?
**Answer:**  
No. It returns a Unix timestamp.

### 3. Is `json.dumps()` the same as `json.dump()`?
**Answer:**  
No. `dumps()` returns a string, while `dump()` writes to a file.

### 4. Can `re.search()` find more than one match?
**Answer:**  
It returns the first match. Use `findall()` for all matches.

### 5. Is `numpy` always available after installing Python?
**Answer:**  
No. It usually must be installed separately.

### 6. Can `random.choice()` work on strings?
**Answer:**  
Yes. A string is also a sequence.

### 7. Does `os.mkdir()` create nested directories automatically?
**Answer:**  
Not in the same way as `os.makedirs()`; `os.mkdir()` creates a single directory.

### 8. Is CSV the same as JSON?
**Answer:**  
No. CSV is tabular text format; JSON is structured key-value / nested data format.
