# 1. Introduction to Data Types

## What are data types?
A **data type** is a classification that tells Python what kind of value is being stored and what kind of operations are allowed on it.

Examples:
- whole numbers → `int`
- decimal numbers → `float`
- text → `str`
- true/false values → `bool`

## Why data types matter
Data types are important because they:

- help Python understand how to store and process data
- determine which operations are valid
- improve program correctness
- reduce certain types of errors
- make code easier to understand


## Common built-in data types introduced here

### Basic data types

- `float`

- `bool`

### Advanced data types (mentioned for later study)

# 2. Integer Data Type (`int`)

An integer is a whole number without a decimal point.

Examples:
- `-10`
- `0`
- `35`
- `1000`


````python
age = 35
print(age)
print(type(age))
````

### Key points
- integers represent whole numbers
- they can be positive, negative, or zero

# 3. Floating Point Data Type (`float`)

A float is a number with a decimal point.

Examples:
- `5.11`
- `3.14`
- `0.5`
- `-7.25`


````python
height = 5.11
print(height)
print(type(height))
````

### Key points
- floats store decimal values

# 4. String Data Type (`str`)

A string is a sequence of characters enclosed in quotes.

Examples:
- `"Krish"`
- `'Hello'`
- `"Python123"`


````python
name = "Krish"
print(name)
print(type(name))
````

### Key points
- strings represent text
- they can use single quotes or double quotes

  - `.upper()`
  - `.lower()`
  - `.strip()`

  - `.split()`


````python
text = "hello world"
print(text.upper())
print(text.title())
print(text.replace("world", "Python"))
````

# 5. Boolean Data Type (`bool`)

Booleans represent truth values:

- `True`
- `False`

These are heavily used in:
- conditions
- comparisons
- decision making
- logical expressions


````python
is_student = True
print(is_student)
print(type(is_student))
````

## Boolean values from comparisons
Comparison expressions often return booleans.


````python
a = 10
b = 10
print(a == b)
print(type(a == b))
````

## Using `bool()`
The `bool()` function converts values into `True` or `False`.


````python
print(bool())       # default -> False
print(bool(0))      # False
print(bool(1))      # True
print(bool(""))     # False
print(bool("Hi"))   # True
````

# 6. Type Checking

Use the `type()` function to check the data type of a value or variable.


````python
age = 35
height = 5.11
name = "Krish"
is_active = True

print(type(age))
print(type(height))
print(type(name))
print(type(is_active))
````

# 7. Type Conversion (Type Casting)

Type conversion means converting one data type into another.

Common conversion functions:
- `int()`
- `float()`
- `str()`
- `bool()`


## Examples of type conversion


````python
# int to str
x = 5
x_str = str(x)
print(x_str, type(x_str))

# str to int
num_text = "25"
num_int = int(num_text)
print(num_int, type(num_int))

# int to float
n = 10
n_float = float(n)
print(n_float, type(n_float))

# float to int
pi = 3.99
pi_int = int(pi)
print(pi_int, type(pi_int))
````

### Important notes
- `int(3.99)` becomes `3` because the decimal part is truncated
- `int("25")` works
- `int("25.5")` fails with `ValueError`
- `int("Krish")` fails with `ValueError`

# 11. Arithmetic Operators

Arithmetic operators are used for mathematical calculations.

Let:
```python
a = 10
b = 15
```


````python
a = 10
b = 15

add_result = a + b
sub_result = a - b
mul_result = a * b
div_result = a / b
floor_div_result = a // b
mod_result = a % b
exp_result = a ** b

print("Addition:", add_result)
print("Subtraction:", sub_result)
print("Multiplication:", mul_result)
print("Division:", div_result)
print("Floor division:", floor_div_result)
print("Modulus:", mod_result)
print("Exponentiation:", exp_result)
````

## Meaning of each arithmetic operator

### `+` Addition
Adds two values.

Example:
```python
10 + 15 = 25
```

### `-` Subtraction
Subtracts right operand from left operand.

Example:
```python
10 - 15 = -5
```

### `*` Multiplication
Multiplies two values.

Example:
```python
10 * 15 = 150
```

### `/` Division
Performs normal division and returns a float.

Example:
```python
10 / 15 = 0.666...
```

### `//` Floor Division
Divides and returns the floor of the result.

Example:
```python
10 // 15 = 0
21 // 5 = 4
```

### `%` Modulus
Returns the remainder after division.

Example:
```python
10 % 15 = 10
21 % 5 = 1
```

### `**` Exponentiation
Raises left operand to the power of the right operand.

Example:
```python
2 ** 3 = 8
```


````python
print("21 / 5 =", 21 / 5)
print("21 // 5 =", 21 // 5)
print("21 % 5 =", 21 % 5)
````

## Division vs Floor Division

### Normal division `/`
Returns the exact decimal result.

### Floor division `//`
Returns the floor value of the division result.

Example:
- `21 / 5 = 4.2`
- `21 // 5 = 4`

For negative values:
- `-21 // 5 = -5`
because floor means going to the next lower integer.


````python
print(-21 / 5)
print(-21 // 5)
````

# 12. Comparison Operators

Comparison operators compare two values and return `True` or `False`.


````python
a = 10
b = 10

print(a == b)   # equal to
print(a != b)   # not equal to
print(a > b)    # greater than
print(a < b)    # less than
print(a >= b)   # greater than or equal to
print(a <= b)   # less than or equal to
````

## Comparison operators explained

### `==` Equal to
Checks whether two values are equal.

### `!=` Not equal to
Checks whether two values are different.

### `>` Greater than
Checks whether left value is greater than right value.

### `<` Less than
Checks whether left value is smaller than right value.

### `>=` Greater than or equal to
Checks whether left value is either greater than or equal to right value.

### `<=` Less than or equal to
Checks whether left value is either less than or equal to right value.


## Comparison with strings
Strings are also comparable.

Comparison is case-sensitive.


````python
str1 = "Krish"
str2 = "Krish"
str3 = "krish"

print(str1 == str2)   # True
print(str1 == str3)   # False
print(str1 != str3)   # True
````

### Important point
String comparison is case-sensitive, so:
- `"Krish"` is not equal to `"krish"`


# 13. Logical Operators

Logical operators combine or invert boolean expressions.

Main logical operators:
- `and`
- `or`
- `not`


## `and` operator
Returns `True` only when **both conditions are True**.


````python
x = True
y = True
print(x and y)

y = False
print(x and y)
````

Truth table for `and`:

| x | y | x and y |
|---|---|---------|
| True | True | True |
| True | False | False |
| False | True | False |
| False | False | False |


## `or` operator
Returns `False` only when **both conditions are False**.


````python
x = True
y = False
print(x or y)

x = False
y = False
print(x or y)
````

Truth table for `or`:

| x | y | x or y |
|---|---|--------|
| True | True | True |
| True | False | True |
| False | True | True |
| False | False | False |


## `not` operator
Reverses a boolean value.

- `not True` → `False`
- `not False` → `True`


````python
x = True
print(not x)

x = False
print(not x)
````

# 14. Combined Examples

Logical operators are often used with comparisons.


````python
age = 25
salary = 50000

print(age > 18 and salary > 30000)
print(age > 30 or salary > 30000)
print(not(age > 18))
````

# 19. Interview Concepts and Model Answers

### 1. What are data types in Python?
**Model answer:**  
Data types classify the kind of data a variable holds, such as integer, float, string, or boolean, and determine what operations can be performed on that data.

### 2. Why are data types important?
**Model answer:**  
They help Python handle values correctly, support valid operations, improve clarity, and reduce certain classes of programming errors.

### 3. Explain the difference between `int`, `float`, `str`, and `bool`.
**Model answer:**  
`int` stores whole numbers, `float` stores decimal values, `str` stores text, and `bool` stores truth values `True` and `False`.

### 4. What is type casting in Python?
**Model answer:**  
Type casting is the process of converting one data type into another using functions like `int()`, `float()`, `str()`, and `bool()`.

### 5. What is the difference between arithmetic and comparison operators?
**Model answer:**  
Arithmetic operators perform mathematical calculations, while comparison operators compare values and return boolean results.

### 6. What is floor division?
**Model answer:**  
Floor division uses `//` and returns the floor of the quotient instead of the exact decimal value.

### 7. What is the purpose of the modulus operator?
**Model answer:**  
The modulus operator `%` returns the remainder after division and is often used to check divisibility or cycle through values.

### 8. Explain logical operators in Python.
**Model answer:**  
Logical operators combine or invert boolean expressions. `and` requires both operands to be true, `or` requires at least one true operand, and `not` reverses the truth value.

### 9. Why does `"hello" + 5` fail?
**Model answer:**  
Because Python does not allow direct addition of a string and an integer. The integer must first be converted to a string.

### 10. What type does `input()` return, and why does that matter?
**Model answer:**  
`input()` returns a string by default. If numeric calculations are needed, the input must be converted using `int()` or `float()`.
