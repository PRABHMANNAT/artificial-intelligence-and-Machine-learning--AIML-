# 1. Introduction to Inheritance

Inheritance is a fundamental concept in object-oriented programming.

It allows one class to inherit attributes and methods from another class.

## Why inheritance matters
Inheritance helps with:
- code reuse
- cleaner design
- avoiding duplication
- modeling real-world relationships

## Real-world analogy
A child may inherit properties or characteristics from parents.

Similarly in Python:
- parent class = base class
- child class = derived class


# 2. Parent Class and Child Class

## Parent class
The class whose features are inherited.

## Child class
The class that inherits from the parent and can also add its own features.

## Simple idea
If a child class inherits from a parent class, it can use:
- parent attributes
- parent methods

without rewriting them from scratch.


# 3. Single Inheritance

Single inheritance means:
> one child class inherits from one parent class


## Example: Parent class `Car`


````python
class Car:
    def __init__(self, windows, doors, engine_type):
        self.windows = windows
        self.doors = doors
        self.engine_type = engine_type

    def drive(self):
        print(f"The person will drive the {self.engine_type} car.")


car1 = Car(4, 5, "petrol")
car1.drive()

print("Windows:", car1.windows)
print("Doors:", car1.doors)
print("Engine type:", car1.engine_type)
````

## Explanation
The `Car` class has:
- attributes:
  - `windows`
  - `doors`
  - `engine_type`
- method:
  - `drive()`

An object of `Car` gets these values when it is created.


# 4. Child Class Example: `Tesla`

Now let us create a child class `Tesla` that inherits from `Car`.

Tesla should inherit:
- windows
- doors
- engine type
- drive method

and also add:
- `is_self_driving`
- `self_driving()` method


````python
class Car:
    def __init__(self, windows, doors, engine_type):
        self.windows = windows
        self.doors = doors
        self.engine_type = engine_type

    def drive(self):
        print(f"The person will drive the {self.engine_type} car.")


class Tesla(Car):
    def __init__(self, windows, doors, engine_type, is_self_driving):
        super().__init__(windows, doors, engine_type)
        self.is_self_driving = is_self_driving

    def self_driving(self):
        print(f"Tesla supports self driving: {self.is_self_driving}")


tesla1 = Tesla(4, 5, "electric", True)
tesla1.drive()
tesla1.self_driving()

print("Windows:", tesla1.windows)
print("Doors:", tesla1.doors)
print("Engine type:", tesla1.engine_type)
print("Self driving:", tesla1.is_self_driving)
````

## Explanation
`Tesla(Car)` means:
- Tesla inherits from Car

Inside Tesla's constructor:
```python
super().__init__(windows, doors, engine_type)
```
calls the constructor of the parent class `Car`.

So Tesla gets:
- `windows`
- `doors`
- `engine_type`

Then Tesla adds its own attribute:
- `is_self_driving`


# 5. What Does `super()` Do?

`super()` is used to refer to the parent class in a clean and maintainable way.

## Example
```python
super().__init__(windows, doors, engine_type)
```

This tells Python:
> call the parent class constructor and initialize the inherited attributes


## Why `super()` is useful
- avoids repeating parent class name
- cleaner and safer
- important in larger inheritance hierarchies
- especially useful with multiple inheritance


# 6. Inherited Methods

A child class can directly use parent methods if they are inherited.

In the Tesla example:
```python
tesla1.drive()
```
works even though `drive()` is defined in `Car`, not in `Tesla`.


## Example


````python
class Car:
    def __init__(self, windows, doors, engine_type):
        self.windows = windows
        self.doors = doors
        self.engine_type = engine_type

    def drive(self):
        return f"The person will drive the {self.engine_type} car."


class Tesla(Car):
    def __init__(self, windows, doors, engine_type, is_self_driving):
        super().__init__(windows, doors, engine_type)
        self.is_self_driving = is_self_driving


tesla = Tesla(4, 5, "electric", True)
print(tesla.drive())
````

# 7. Method Overriding

If a child class defines a method with the same name as the parent method, the child version overrides the parent version.


````python
class Car:
    def drive(self):
        return "Driving a normal car"


class Tesla(Car):
    def drive(self):
        return "Driving an electric Tesla"


car = Car()
tesla = Tesla()

print(car.drive())
print(tesla.drive())
````

## Explanation
Even though `Tesla` inherits from `Car`, it defines its own `drive()` method.

So when `tesla.drive()` is called, Python uses the Tesla version.


# 8. Multiple Inheritance

Multiple inheritance means:
> one child class inherits from more than one parent class


## Example structure
```python
class Dog(Animal, Pet):
    ...
```

Here:
- `Animal` is one parent class
- `Pet` is another parent class
- `Dog` inherits from both


## Example: Multiple inheritance with `Animal` and `Pet`


````python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "Subclasses must implement this method."


class Pet:
    def __init__(self, owner):
        self.owner = owner


class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)

    def speak(self):
        return f"{self.name} says woof"


dog = Dog("Buddy", "Krish")
print(dog.speak())
print("Owner:", dog.owner)
print("Name:", dog.name)
````

## Explanation
`Dog` inherits from both:
- `Animal`
- `Pet`

From `Animal`, it gets:
- `name`
- `speak()` concept

From `Pet`, it gets:
- `owner`

Then `Dog` defines its own `speak()` method.


# 9. Direct Parent Constructor Calls vs `super()`

In the multiple inheritance example, the transcript used:

```python
Animal.__init__(self, name)
Pet.__init__(self, owner)
```

This works for simple teaching examples.

But in bigger designs, Python usually encourages cooperative inheritance with `super()`.

For beginners:
- direct parent calls are okay to understand the concept
- `super()` becomes more important later


# 10. Method Resolution Order (MRO)

In multiple inheritance, Python must decide the order in which parent classes are searched.

This order is called **Method Resolution Order**.


````python
class A:
    def show(self):
        return "A"


class B(A):
    pass


class C(A):
    def show(self):
        return "C"


class D(B, C):
    pass


d = D()
print(d.show())
print(D.__mro__)
````

## Explanation
Python checks classes in MRO order to decide:
- which method to use
- which parent gets priority

This becomes very important in multiple inheritance.

## Multiple inheritance
One child inherits from more than one parent.

Example:

# 12. Why Inheritance Is Useful

Inheritance is useful because it helps:
- avoid writing the same code repeatedly
- keep common logic in one base class
- build specialized child classes from a common blueprint
- model real-world hierarchy


## Example idea
A base class `Vehicle` could provide:
- wheels
- engine
- move()

Child classes could be:
- `Car`
- `Bike`
- `Truck`

Each child reuses the common features and adds its own behavior.


# 13. Common Errors in Inheritance

## 1. Forgetting to call parent constructor
If the parent has important attributes, not calling the parent constructor can leave the object incomplete.

## 2. Using wrong attribute names
If an attribute is inherited, it should be referenced correctly with `self.attribute_name`.

## 3. Confusing inherited methods with overridden methods
If the child defines the same method name, the child version is used.

## 4. Mismanaging multiple inheritance
Multiple inheritance can get confusing if class relationships are not designed carefully.

## 5. Printing a method that already prints
If a method uses `print()` but does not return anything, wrapping it with `print(...)` can show an extra `None`.


# 14. Best Practices

## 1. Use inheritance only when there is a genuine “is-a” relationship
Example:
- Tesla **is a** Car
- Dog **is an** Animal

## 2. Keep base classes general
Put common behavior in the parent class.

## 3. Keep child classes specialized
Add only the extra child-specific features.

## 4. Use `super()` when appropriate
Especially helpful in clean inheritance design.

## 5. Be careful with multiple inheritance
Use it only when it truly makes sense.

## 6. Prefer readability
Inheritance is powerful, but clarity matters more than clever design.

# 17. Tricky Questions

**Answer:**  
Not literally. The child class gains access to parent behavior through the inheritance relationship.

### 2. If both parent and child have a method with the same name, which one runs?
**Answer:**  
The child version runs.

### 3. Can a child class have extra attributes not present in the parent class?
**Answer:**  
Yes.

### 4. Can Python support more than one parent class?
**Answer:**  
Yes, through multiple inheritance.

### 5. Is `super()` mandatory?
**Answer:**  
No, but it is often the cleaner and recommended approach.

### 6. Can multiple inheritance become confusing?
**Answer:**  
Yes, especially because of method resolution order.

### 7. Is inheritance always the best design choice?
**Answer:**  
No. Sometimes composition is better than inheritance.

### 8. If a parent method only prints and does not return, what happens if you wrap it in `print()`?
**Answer:**  
You may see the printed message and then `None`.
