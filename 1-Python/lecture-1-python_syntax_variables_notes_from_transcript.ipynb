# 1. Syntax vs Semantics

## What is Syntax?
**Syntax** means the set of rules that define how code must be written so Python can understand it.

In simple words:
> Syntax is the **structure/form** of code.

Examples of syntax-related rules:
- proper indentation
- correct spelling of keywords
- matching parentheses
- valid variable names
- using `:` after `if`, `for`, `while`, `def`, etc.

## What is Semantics?
**Semantics** means the **meaning** of the code — what the code is supposed to do when it runs.

In simple words:
> Syntax is how code is written, semantics is what the code means.

### Example
```python
print("Hello")
```

- Syntax: valid
- Semantics: prints `Hello`

Another example:
```python
x = 10
y = 20
print(x + y)
```

- Syntax: valid
- Semantics: adds the two numbers and prints `30`


## Syntax Error vs Semantic/Logical Error

### Syntax Error
Python cannot even parse the code.

Example:
```python
if True
    print("Hello")
```

Problem: missing `:` after `True`

### Semantic / Logical Error
The code runs, but the result is not what you intended.

Example:
```python
price = 100
discount = 10
final = price - 100   # logically wrong if you intended 10%
```

This code is syntactically valid, but semantically/logically wrong for the intended task.


# 2. Comments in Python

## Single-line comments
Use `#` for a single-line comment.

Example:
```python
# This is a comment
print("Hello World")
```

## “Multiline comments”
Python has **no separate multiline comment syntax** like some languages.

Instead, people often use:
- many `#` lines, or
- triple-quoted strings (`'''...'''` or `"""..."""`) as notes/docstrings

Example:
```python
"""
This is a multiline string.
It is often used like a note or docstring.
"""
```

### Best practice
For real comments in code:
- prefer `#` for explanations
- use triple quotes mainly for **docstrings**


````python
# Single-line comment example
print("Hello World")  # inline comment

# A multiline string (not a true comment)
"""
This is a multiline string.
People often use it like a comment,
but technically it is a string object.
"""
````

# 3. Basic Syntax Rules in Python

## 3.1 Case Sensitivity
Python is **case-sensitive**.

That means:
- `name`
- `Name`
- `NAME`

are all different identifiers.


````python
name = "Krish"
Name = "Nick"

print(name)
print(Name)
print(name == Name)
````

### Key idea
Because Python is case-sensitive:
- variable names must be used with the exact same spelling and casing
- `age`, `Age`, and `AGE` are different


## 3.2 Indentation
Python uses **indentation** to define blocks of code.

Unlike languages such as C, C++, or Java that often use `{ }`, Python uses spaces (commonly 4 spaces) or tabs to indicate a block.

### Where indentation matters
Indentation is required after statements like:
- `if`
- `elif`
- `else`
- `for`
- `while`
- `def`
- `class`
- `try`
- `except`


````python
age = 32

if age >= 30:
    print("Inside if block:", age)

print("Outside if block:", age)
````

### Why indentation matters
The indented code belongs to the block above it.

In this example:
```python
if age >= 30:
    print(age)
```

`print(age)` runs only when the condition is true.


### Indentation Error
If indentation is missing where required, Python raises an `IndentationError`.


````python
# Uncomment to see the error

# age = 32
# if age >= 30:
# print(age)
````

## 3.3 Line Continuation
Long statements can continue onto the next line.

### Method 1: Backslash
```python
total = 1 + 2 + 3 + \
        4 + 5 + 6
```

### Method 2: Preferred modern style
Inside parentheses, brackets, or braces, Python allows implicit continuation:
```python
total = (
    1 + 2 + 3 +
    4 + 5 + 6
)
```

The second style is usually cleaner and safer.


````python
# Using backslash
total1 = 1 + 2 + 3 + \
         4 + 5 + 6

# Using parentheses (preferred)
total2 = (
    1 + 2 + 3 +
    4 + 5 + 6
)

print(total1, total2)
````

## 3.4 Multiple Statements on One Line
You *can* write multiple statements on a single line using `;`

Example:
```python
x = 5; y = 10; z = x + y
```

But in real projects, this is usually avoided for readability.


````python
x = 5; y = 10; z = x + y
print(z)
````

# 5. Variables in Python

## What is a variable?
A variable is a name that refers to a value stored in memory.

You can think of it as a **label** attached to some data.

Example:
```python
age = 32
name = "Krish"
```

Here:
- `age` stores an integer
- `name` stores a string


## How variables are created
In Python, variables are created **when you assign a value** to them.

You do not have to explicitly declare them first like in some other languages.


````python
a = 100
print(a)
````

## Declaring and assigning variables
Examples:
```python
age = 32
height = 6.1
name = "Krish"
is_student = True
```


````python
age = 32
height = 6.1
name = "Krish"
is_student = True

print("Age:", age)
print("Height:", height)
print("Name:", name)
print("Is student:", is_student)
````

# 6. Naming Conventions for Variables

## Rules
A variable name:
- must start with a letter or underscore `_`
- can contain letters, numbers, and underscores
- cannot start with a number
- cannot contain spaces
- cannot contain operators/symbols like `-`, `@`, `%`
- is case-sensitive

## Good examples
- `first_name`
- `last_name`
- `student_age`
- `_private_value`

## Bad examples
- `2age`
- `first-name`
- `@name`
- `my value`


````python
# Valid variable names
first_name = "Krish"
last_name = "Nick"
student_age_1 = 25
_private_value = 100

print(first_name, last_name, student_age_1, _private_value)
````

````python
# Invalid examples (kept commented so the notebook runs)

# 2age = 30          # invalid: starts with a number
# first-name = "A"   # invalid: '-' is treated as subtraction
# @name = "Krish"    # invalid syntax
# my value = 10      # invalid: spaces not allowed
````

## Recommended style
Use **snake_case** for normal variable names in Python.

Example:
```python
total_marks = 95
student_name = "Riya"
```


# 7. Variable Types and Dynamic Typing

Python is **dynamically typed**.

This means:
- you do not declare the type explicitly
- Python determines the type at runtime

Common types:
- `int`
- `float`
- `str`
- `bool`


````python
age = 25          # int
height = 6.1      # float
name = "Krish"    # str
is_student = True # bool

print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
````

## Type inference
Python automatically infers the type from the assigned value.

Example:
```python
age = 25
```

Python understands that `age` is an integer because the assigned value is an integer.


## Dynamic typing example
The same variable can hold different types at different times.


````python
var = 10
print(var, type(var))

var = "Hello"
print(var, type(var))

var = 3.14
print(var, type(var))
````

### Important note
Python allows the variable to point to a new value of a different type, but this should be used carefully in real code to avoid confusion.


# 8. Type Checking and Type Conversion

## Type checking
Use `type()` to check a variable's type.

Example:

````python
age = 25
print(type(age))
````

## Type conversion (type casting)

You can convert values between types using:
- `int()`
- `float()`
- `str()`
- `bool()`

### Examples
- integer to string
- string containing digits to integer
- float to integer


````python
age = 25
age_str = str(age)

print(age_str)
print(type(age_str))
````

````python
num_text = "25"
num_int = int(num_text)

print(num_int)
print(type(num_int))
````

````python
height = 5.11
height_int = int(height)

print(height_int)
print(type(height_int))
````

### Important behavior of `int()` with float
```python
int(5.11)
```
returns `5`, because Python **truncates** the decimal part. It does **not** round.


## Invalid conversion example
Not every string can be converted to an integer.

Example:
```python

```

This raises:
```python
ValueError
```
because `"Krish"` is not a numeric string.

# 11. Quick Revision / Cheatsheet

## Syntax
- Syntax = rules for writing correct code
- Semantics = meaning of the code
- Python is case-sensitive
- Use indentation to define blocks
- Use `#` for single-line comments
- Triple quotes create multiline strings/docstrings
- Use `;` for multiple statements on one line (rarely recommended)
- Use `\` or parentheses for line continuation

## Variables
- Variables store values
- Created at assignment time
- No explicit type declaration needed
- Python is dynamically typed
- Common types: `int`, `float`, `str`, `bool`
- Use `type()` to inspect data type

## Naming Rules
- Start with letter or `_`
- Can contain letters, numbers, `_`
- Cannot start with a number
- Cannot contain spaces or symbols like `@` or `-`
- Use descriptive names
- Prefer `snake_case`

## Type Conversion
- `str(25)` → `"25"`
- `int("25")` → `25`
- `float(5)` → `5.0`
- `int(5.99)` → `5`

## Common Errors
- `SyntaxError` → invalid structure
- `IndentationError` → wrong/missing indentation
- `NameError` → variable not defined
- `ValueError` → invalid conversion like `int("abc")`
