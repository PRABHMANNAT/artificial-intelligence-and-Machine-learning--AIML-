# 1. Introduction to Operator Overloading

Operator overloading means customizing how operators behave for your own objects.

## Why it matters
Normally operators like:
- `+`
- `-`
- `*`
- `/`
- `==`
- `<`
- `>`

work on built-in types such as integers, floats, and strings.

With operator overloading, you can make these operators work naturally on your own custom classes.


# 2. Operator Overloading Uses Magic Methods

Operator overloading is done using magic methods (dunder methods).

## Common operator magic methods
- `__add__` → `+`
- `__sub__` → `-`
- `__mul__` → `*`
- `__truediv__` → `/`
- `__eq__` → `==`
- `__lt__` → `<`
- `__gt__` → `>`

These methods let you define what those operators should do for your custom objects.


# 3. Vector Class Example

A clean way to understand operator overloading is with vectors.

We will create a class with:
- `x`
- `y`

and overload operators so that vectors can be added, subtracted, compared, and multiplied by scalars.


````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"


v1 = Vector(2, 3)
v2 = Vector(4, 5)

print("v1:", v1)
print("v2:", v2)
print("v1 + v2:", v1 + v2)
print("v1 - v2:", v1 - v2)
print("v1 * 3:", v1 * 3)
print("v1 == v2:", v1 == v2)
print("Vector(2, 3) == Vector(2, 3):", Vector(2, 3) == Vector(2, 3))
````

## Explanation

### `__add__`
This defines what happens when:
```python
v1 + v2
```

### `__sub__`
This defines what happens when:
```python
v1 - v2
```

### `__mul__`
This defines what happens when:
```python
v1 * 3
```

### `__eq__`
This defines what happens when:
```python
v1 == v2
```

### `__repr__`
This defines how the object is displayed in a clearer developer-friendly form.

# 5. Overloading Comparison Operators

You can also overload comparison operators.

## Example: Compare vectors by magnitude


````python
class VectorWithMagnitude:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def magnitude_squared(self):
        return self.x**2 + self.y**2

    def __lt__(self, other):
        return self.magnitude_squared() < other.magnitude_squared()

    def __gt__(self, other):
        return self.magnitude_squared() > other.magnitude_squared()

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    def __repr__(self):
        return f"VectorWithMagnitude({self.x}, {self.y})"


a = VectorWithMagnitude(1, 2)
b = VectorWithMagnitude(3, 4)

print("a:", a)
print("b:", b)
print("a < b:", a < b)
print("a > b:", a > b)
print("a == b:", a == b)
````

## Explanation
This example compares vectors using squared magnitude:
```python
x^2 + y^2
```

So now `<` and `>` behave according to vector size, not coordinate-by-coordinate comparison.


# 6. Another Example: Complex-Number Style Class

The transcript suggested exploring operator overloading for complex numbers. Here is a clean teaching version.


````python
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __add__(self, other):
        return ComplexNumber(self.real + other.real, self.imag + other.imag)

    def __sub__(self, other):
        return ComplexNumber(self.real - other.real, self.imag - other.imag)

    def __mul__(self, other):
        real_part = self.real * other.real - self.imag * other.imag
        imag_part = self.real * other.imag + self.imag * other.real
        return ComplexNumber(real_part, imag_part)

    def __repr__(self):
        sign = "+" if self.imag >= 0 else "-"
        return f"{self.real} {sign} {abs(self.imag)}i"


c1 = ComplexNumber(2, 3)
c2 = ComplexNumber(4, 5)

print("c1:", c1)
print("c2:", c2)
print("c1 + c2:", c1 + c2)
print("c1 - c2:", c1 - c2)
print("c1 * c2:", c1 * c2)
````

# 7. Why Operator Overloading Is Useful

Operator overloading is useful because it lets your custom objects behave naturally.

Instead of writing:
```python
v1.add(v2)
```

you can write:
```python
v1 + v2
```

That makes code:
- cleaner
- more expressive
- closer to math or domain logic


# 8. Best Practices for Operator Overloading

## Good use cases
Use operator overloading when the operator meaning is natural for the object.

Examples:
- vectors
- matrices
- complex numbers
- fractions
- measurements

## Avoid confusing behavior
Do not overload operators in surprising ways.

Example:
- `+` should usually mean some kind of combination or addition
- `==` should mean equality


# 9. Common Errors in Operator Overloading

## 1. Forgetting to return a new object
Your magic method usually should return a meaningful result.

## 2. Comparing objects incorrectly
For equality, make sure all necessary fields are checked.

## 3. Using unsupported operand types
If your `__mul__` expects a scalar but gets another object, you may need extra checks.

## 4. Not defining `__repr__` / `__str__`
Without them, printed objects look less readable.

## 5. Overloading operators in unnatural ways
This hurts readability.


# 10. Introduction to Custom Exceptions

## Why custom exceptions matter
Custom exceptions are useful when you want to:
- enforce business rules
- raise meaningful application-specific errors
- separate domain errors from generic built-in errors
- show clearer messages to users or developers

# 12. Example Scenario: Exam Form Validation

Suppose an exam allows only candidates whose age is between 20 and 30.

If age falls outside that range, we want to raise a custom exception.


````python
from datetime import datetime

class Error(Exception):
    pass


class DOBException(Error):
    pass


def validate_dob_year(year_of_birth):
    current_year = datetime.now().year
    age = current_year - year_of_birth

    if 20 <= age <= 30:
        return f"Age is valid ({age}). You can apply for the exam."
    else:
        raise DOBException(f"Age {age} is not in the allowed range 20 to 30.")


# Valid example
print(validate_dob_year(datetime.now().year - 25))
````

# 13. Handling the Custom Exception with `try` / `except`

If you raise an exception, you should usually handle it where appropriate.


````python
from datetime import datetime

class Error(Exception):
    pass


class DOBException(Error):
    pass


try:
    current_year = datetime.now().year
    year_of_birth = current_year - 37  # makes age 37
    result = validate_dob_year(year_of_birth)
    print(result)
except DOBException as e:
    print("Custom exception caught:", e)
````

## Explanation
- `raise DOBException(...)` creates and throws the custom exception
- `except DOBException as e` catches it
- the message inside the exception is available in `e`


# 14. Interactive Version of the DOB Example

This is a cleaned version of the transcript logic.


````python
from datetime import datetime

class Error(Exception):
    pass


class DOBException(Error):
    pass


def check_exam_eligibility(year_of_birth):
    current_year = datetime.now().year
    age = current_year - year_of_birth

    if 20 <= age <= 30:
        return f"Age is valid ({age}). You can apply for the exam."
    else:
        raise DOBException("Your age should be between 20 and 30 for this exam.")


# Example runs:
for yob in [datetime.now().year - 22, datetime.now().year - 35]:
    try:
        print(f"Checking year of birth: {yob}")
        print(check_exam_eligibility(yob))
    except DOBException as e:
        print("Error:", e)
    print("-" * 50)
````

# 15. Why Use a Custom Exception Instead of a Normal `ValueError`?

You *could* use built-in exceptions such as `ValueError`, but custom exceptions make intent clearer.

## Example
- `ValueError` → generic problem with a value
- `DOBException` → very specific business rule failure

This makes debugging and app design cleaner.


# 16. Custom Exception with Extra Information

Custom exceptions can also store extra details.


````python
class AgeRangeException(Exception):
    def __init__(self, age, min_age=20, max_age=30):
        self.age = age
        self.min_age = min_age
        self.max_age = max_age
        super().__init__(f"Age {age} is not allowed. Allowed range: {min_age} to {max_age}.")


def validate_age(age):
    if not (20 <= age <= 30):
        raise AgeRangeException(age)
    return "Valid age"


try:
    print(validate_age(19))
except AgeRangeException as e:
    print(e)
    print("Age received:", e.age)
````

# 17. Common Exception Flow

The normal flow is:

1. detect an invalid condition
2. `raise` an exception
3. catch it with `except` if you want to handle it
4. show a helpful message or take corrective action

# 19. Common Errors in Custom Exceptions

## 1. Raising without catching where needed
This may crash the program if not handled upstream.

## 2. Using unclear exception names
Names should describe the domain problem.

## 3. Overusing generic custom exceptions
Too many vague exception classes can make code harder to understand.

## 4. Hardcoding values unnecessarily
For age validation, using the current year dynamically is better than hardcoding a year.

## 5. Using custom exceptions for normal control flow too often
Exceptions should represent exceptional or rule-breaking conditions, not everyday branching.
