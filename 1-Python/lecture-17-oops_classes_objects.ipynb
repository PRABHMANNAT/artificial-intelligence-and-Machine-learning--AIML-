## 2) What Is OOP?

**Object-Oriented Programming (OOP)** is a programming paradigm that uses **objects** to design programs.

Instead of writing everything as separate variables and functions, we group related:
- **data** → attributes / properties
- **behavior** → methods

inside a **class**.

### Real-world analogy
Think of a class as a **blueprint**.

Example:
- `Car` is a blueprint
- Audi, BMW, Tesla are actual objects created from that blueprint


## 3) What Is a Class?

A **class** is a blueprint for creating objects.

A class can contain:
- **attributes** → data / properties
- **methods** → functions that define behavior

### Example
```python
class Car:
    pass
```

This creates an empty class named `Car`.


````python
class Car:
    pass
````

## 4) What Is an Object?

An **object** is an instance of a class.

If `Car` is a class, then:

```python
audi = Car()
bmw = Car()
```

Both `audi` and `bmw` are objects created from the `Car` class.

### Key idea
- class = blueprint
- object = real instance created from the blueprint


````python
class Car:
    pass

audi = Car()
bmw = Car()

print(type(audi))
print(audi)
print(bmw)
````

## 7) Why Setting Attributes Randomly Is Not the Best Approach

The transcript first shows something like:

```python
audi = Car()
audi.windows = 4

tata = Car()
tata.doors = 4
```

This works, but it is not a good design because different objects may end up with inconsistent attributes.

Example:

```python
print(tata.windows)
```

This may fail because `windows` was never assigned to `tata`.

### Problem
Objects created from the same class should usually have a consistent structure.

### Better approach
Define the expected attributes using a constructor: `__init__()`


````python
class Car:
    pass

audi = Car()
audi.windows = 4

tata = Car()
tata.doors = 4

print(audi.windows)
# print(tata.windows)  # This would raise AttributeError
````

## 8) The Constructor: `__init__()`

The **constructor** in Python is the special method:

```python
def __init__(self, ...):
```

It runs automatically when an object is created.

### Why use it?
It helps initialize object attributes properly at object creation time.

### Example
```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

Now every `Dog` object must be created with:
- a name
- an age


````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
````

## 9) What Is `self`?

`self` refers to the **current object**.

It is used to access the instance variables and methods of that object.

### Example
```python
self.name = name
```

Here:
- `self.name` → instance variable stored in the object
- `name` → value passed during object creation

### Important note
Technically, the first parameter does not have to be named `self`, but using `self` is the standard Python convention.


## 10) Creating Objects with a Constructor

Once the constructor is defined, object creation must follow its parameters.

### Example
```python
dog1 = Dog("Buddy", 3)
dog2 = Dog("Lucy", 4)
```

Now:
- `dog1.name` is `"Buddy"`
- `dog1.age` is `3`
- `dog2.name` is `"Lucy"`
- `dog2.age` is `4`


````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

dog1 = Dog("Buddy", 3)
dog2 = Dog("Lucy", 4)

print(dog1.name)
print(dog1.age)
print(dog2.name)
print(dog2.age)
````

## 11) What Happens If Constructor Arguments Are Missing?

If the class expects arguments but you do not provide them:

```python
dog = Dog()
```

Python raises an error like:

```python
TypeError: __init__() missing required positional arguments
```

This is useful because it forces correct object creation.


## 12) Methods / Instance Methods

Methods are functions defined inside a class.

When a method works with a specific object, it is usually called an **instance method**.

### Example
A dog can have behavior such as:
- bark
- run
- eat

These behaviors can be written as methods.


## 13) Example: Dog Class with an Instance Method

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says woof")
```

### Why `self` is used in `bark()`
Because the method needs access to the current object's attributes, such as `self.name`.


````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says woof")

dog1 = Dog("Buddy", 3)
dog2 = Dog("Lucy", 4)

dog1.bark()
dog2.bark()
````

## 14) Why Methods Are Powerful

The same class method can behave differently depending on the object.

Example:
- `dog1.bark()` → Buddy says woof
- `dog2.bark()` → Lucy says woof

Same method, different object data.

This is one major strength of OOP.

## 17) Deposit Method

Deposit increases the account balance.

### Example
```python
def deposit(self, amount):
    self.balance += amount
    print(f"{amount} is deposited. New balance is {self.balance}")
```

### Key idea
We use `self.balance` because balance belongs to the object.


## 18) Withdraw Method

Withdraw should reduce the balance only if enough money exists.

### Example logic
```python
def withdraw(self, amount):
    if amount > self.balance:
        print("Insufficient funds")
    else:
        self.balance -= amount
        print(f"{amount} is withdrawn. New balance is {self.balance}")
```

### Why this matters
This adds **real-world business logic** to the class.


## 19) `get_balance()` Method

Sometimes instead of printing directly, a method returns a value.

### Example
```python
def get_balance(self):
    return self.balance
```

Now you can do:
```python
print(account.get_balance())
```

This is often better than directly exposing all internals.


````python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} is deposited. New balance is {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"{amount} is withdrawn. New balance is {self.balance}")

    def get_balance(self):
        return self.balance

account = BankAccount("Krish", 5000)
print(account.balance)

account.deposit(100)
account.withdraw(300)
print(account.get_balance())
````

## 22) Important Corrections / Clarifications from the Transcript

The transcript is spoken naturally, so here are some cleaned-up explanations:

1. **OOP**, not “oops”
   - Informally many teachers say “oops”
   - The formal term is **OOP = Object-Oriented Programming**

2. **`__init__()` is a constructor-like initializer**
   - In Python, people often call it a constructor
   - More precisely, it is the initializer method called after object creation

3. **`self` is not a keyword**
   - It is a naming convention
   - But it should be used as `self` in normal Python code

   - This keeps objects consistent

## 23) Common Mistakes to Avoid

1. **Forgetting `self` in method definitions**
   ```python
   def bark():
       ...
   ```
   This is wrong for instance methods.

2. **Forgetting `self.` before instance variables**
   ```python
   name = name
   ```
   This does not store the value in the object properly.

3. **Creating attributes inconsistently outside the constructor**
   - Makes objects unpredictable

4. **Confusing class and object**
   - `Dog` is a class
   - `dog1 = Dog("Buddy", 3)` is an object

5. **Calling a method without parentheses**
   ```python
   dog1.bark
   ```
   This refers to the method object, but does not execute it.

6. **Using the wrong indentation**
   - Python class and method indentation must be correct

## 28) Tricky / Interview Concepts

### Q1. What is the difference between a class variable and an instance variable?
**Answer:**  
- **Instance variable** belongs to each object separately, like `self.name`
- **Class variable** is shared across all objects of the class

> Note: the transcript mainly covers instance variables.

---

### Q2. Why do we use `self.balance` instead of just `balance` inside methods?
**Answer:**  
Because `self.balance` refers to the object's stored balance.  
Without `self`, Python treats it like a local variable.

---

### Q3. Is `__init__()` the object creator?
**Answer:**  
Strictly speaking, object creation is handled by `__new__()`, while `__init__()` initializes it.  
But for most beginner discussions, `__init__()` is treated as the constructor.

---

### Q4. Why are methods inside classes different from normal functions?
**Answer:**  
Because they usually operate on object data and receive `self`.

---

### Q5. Can one class create many objects?
**Answer:**  
Yes. A single class can create any number of objects.

---

### Q6. Why is OOP considered good for real-world applications?
**Answer:**  
Because real systems naturally contain entities with properties and behaviors.

---

### Q7. What does `dog1.bark()` mean internally?
**Answer:**  
It means Python calls the method with `dog1` passed as `self`.

---

### Q8. Why is setting attributes manually after object creation risky?
**Answer:**  
Because different objects may end up with inconsistent data structure.

---

### Q9. What is the biggest beginner mistake in OOP?
**Answer:**  
Confusing the class definition with the object created from it.

---

### Q10. Why should method names represent behavior?
**Answer:**  
Because methods describe what an object does, such as `deposit`, `withdraw`, or `bark`.

## 30) Best Practices

1. Use **PascalCase** for class names
   - `BankAccount`, `Dog`, `Student`

2. Use `self` consistently

4. Keep method names meaningful
   - `deposit()`, `withdraw()`, `bark()`

5. Model real-world entities cleanly

6. Avoid randomly attaching attributes outside the constructor unless there is a good reason

7. Keep class design simple and focused
