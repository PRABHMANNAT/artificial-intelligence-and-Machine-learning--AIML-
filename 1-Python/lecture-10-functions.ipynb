# 1. Introduction to Functions

A **function** is a reusable block of code that performs a specific task.

## Why functions matter
Functions help with:

- organizing code
- reusing code
- reducing repetition
- improving readability
- making programs easier to test and maintain

## Simple definition
A function takes input (optional), performs work, and may return output.


# 2. General Syntax of a Function

## Basic syntax
```python
def function_name(parameters):
    """Optional docstring"""
    # function body
    return value
```

## Parts of the syntax
- `def` → keyword used to define a function
- `function_name` → the name of the function
- `parameters` → inputs to the function
- `docstring` → optional explanation of what the function does
- function body → the actual code inside the function
- `return` → sends a result back from the function


## Example of an empty function


````python
def empty_function():
    """This is an empty function."""
    pass

print(empty_function)
````

# 3. Why Do We Need Functions?

Without functions, repeated logic must be copied again and again.

That causes:
- duplication
- harder maintenance
- more bugs
- less readable code


## Example without a function


````python
num = 24

if num % 2 == 0:
    print("The number is even")
else:
    print("The number is odd")
````

### Problem
If the same even/odd logic is needed in 20 places, repeating the code is inefficient.

A better solution is to place that logic inside a function and reuse it.


## Same logic using a function


````python
def even_or_odd(num):
    """Check whether a number is even or odd."""
    if num % 2 == 0:
        print("The number is even")
    else:
        print("The number is odd")

even_or_odd(24)
even_or_odd(15)
````

## Why this is better
Now the logic is written once and can be reused many times.


# 4. Defining and Calling a Function

## Function definition
You define a function using `def`.

## Function call
You run it by writing its name followed by parentheses.


````python
def greet_user():
    print("Hello! Welcome.")

greet_user()
````

## Important idea
Defining a function does **not** run it.
It only creates the function.
The function runs when you call it.


````python
def say_hi():
    print("Hi!")

# Function is defined here, but not called yet.
print("Function created.")

say_hi()
````

# 5. Function Parameters

Parameters are variables listed inside the function definition.

They allow the function to accept input.


````python
def greet(name):
    print(f"Hello {name}, welcome to the paradise.")

greet("Krish")
greet("Aman")
````

## Explanation
Here, `name` is a parameter.
When calling the function:
- `"Krish"` is passed to `name`
- `"Aman"` is passed to `name`


## Function with multiple parameters


````python
def add(a, b):
    return a + b

result = add(2, 4)
print(result)
````

### Explanation
The function takes two parameters:
- `a`
- `b`

and returns their sum.


# 6. Return Statement

The `return` statement sends a value back from the function.

## Why `return` is important
It lets you:
- store the result
- reuse the result
- build bigger logic using function outputs


````python
def multiply(a, b):
    return a * b

output = multiply(2, 3)
print(output)
````

## Function without explicit return
If a function does not explicitly return anything, Python returns `None`.


````python
def show_message():
    print("This function prints a message.")

result = show_message()
print("Returned value:", result)
````

# 7. Multiple Parameters Example

Functions can take multiple inputs and work with them.


````python
def calculate_sum(a, b):
    return a + b

print(calculate_sum(10, 20))
print(calculate_sum(-5, 3))
````

# 8. Default Parameters

A default parameter gives a fallback value if no argument is provided.


````python
def greet(name="Guest"):
    print(f"Hello {name}, welcome to the paradise.")

greet()
greet("Krish")
````

## Explanation
- if no value is passed, `name` becomes `"Guest"`
- if a value is passed, it replaces the default


## Important rule
Parameters with default values should usually come after required parameters.


````python
def introduce(name, city="Unknown"):
    print(f"My name is {name} and I live in {city}.")

introduce("Krish")
introduce("Krish", "Delhi")
````

# 9. Positional Arguments

Positional arguments are matched based on the order in which they are passed.


````python
def describe_person(name, age, country):
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(f"Country: {country}")

describe_person("Krish", 32, "India")
````

## Explanation
- first value goes to `name`
- second value goes to `age`
- third value goes to `country`


# 10. Keyword Arguments

Keyword arguments are passed using parameter names.

This makes calls clearer and less dependent on order.


````python
def describe_person(name, age, country):
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(f"Country: {country}")

describe_person(name="Krish", age=32, country="India")
describe_person(country="India", name="Krish", age=32)
````

## Explanation
With keyword arguments, Python matches by name instead of position.


## Positional vs Keyword Arguments
- positional → based on order
- keyword → based on parameter name


## Important rule
Positional arguments must come before keyword arguments in a function call.


````python
def print_details(a, b, c):
    print(a, b, c)

print_details(1, 2, c=3)

# Uncomment to see the error:
# print_details(a=1, 2, 3)
````

# 11. Variable-Length Arguments: `*args`

Sometimes you do not know in advance how many positional arguments will be passed.

Use `*args` for this.


````python
def print_numbers(*args):
    for number in args:
        print(number)

print_numbers(1, 2, 3, 4, 5, 6, "Krish")
````

## Explanation
`*args` collects all positional arguments into a tuple.


````python
def inspect_args(*args):
    print(args)
    print(type(args))

inspect_args(10, 20, 30)
````

## Important note
The name `args` is conventional, but not required.


````python
def show_values(*values):
    print(values)

show_values(1, 2, 3)
````

# 12. Variable-Length Keyword Arguments: `**kwargs`

Sometimes you do not know in advance how many keyword arguments will be passed.

Use `**kwargs` for this.


````python
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(name="Krish", age=32, country="India")
````

## Explanation
`**kwargs` collects keyword arguments into a dictionary.


````python
def inspect_kwargs(**kwargs):
    print(kwargs)
    print(type(kwargs))

inspect_kwargs(name="Krish", age=32)
````

## Important note
The name `kwargs` is also just a convention.


````python
def show_details(**details):
    print(details)

show_details(city="Delhi", pincode=110001)
````

# 13. Combining `*args` and `**kwargs`

A function can accept both positional and keyword arguments.


````python
def print_everything(*args, **kwargs):
    print("Positional arguments:")
    for val in args:
        print(f"  {val}")

    print("Keyword arguments:")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")

print_everything(1, 2, 3, 4, 5, "Krish", age=32, country="India")
````

## Key rule when calling such a function
- positional arguments first
- keyword arguments after that


# 14. Returning Multiple Values

Python functions can return multiple values.

Technically, Python packs them into a tuple.


````python
def return_multiple():
    a = 2
    b = 3
    c = a * b
    return a, b, c

result = return_multiple()
print(result)
print(type(result))
````

## Explanation
Even though it looks like multiple values are returned, Python actually returns a tuple:
```python
(2, 3, 6)
```


## Unpacking multiple returned values


````python
def return_multiple():
    a = 2
    b = 3
    c = a * b
    return a, b, c

x, y, z = return_multiple()
print(x)
print(y)
print(z)
````

# 15. Docstrings

A docstring is a string written inside a function to describe what it does.

It is a very good practice, especially in professional codebases.


````python
def add(a, b):
    """Return the sum of two numbers."""
    return a + b

print(add.__doc__)
````

## Why docstrings matter
They help:
- other developers understand the function
- tools generate documentation
- improve maintainability

# 17. Common Errors with Functions

## 1. Missing indentation
Python requires proper indentation inside function bodies.


````python
# Uncomment to test:
# def bad_function():
# print("Hello")
````

## 2. Missing required positional arguments
If a required parameter is not passed, Python raises `TypeError`.


````python
def greet(name):
    print(f"Hello {name}")

# Uncomment to test:
# greet()
````

## 3. Wrong order of positional and keyword arguments
Positional arguments cannot come after keyword arguments in a function call.


````python
def sample(a, b, c):
    print(a, b, c)

# Uncomment to test:
# sample(a=1, 2, 3)
````

## 4. Forgetting `return`
Sometimes a function prints something but does not return a value.
That can cause confusion when storing results.


````python
def add_and_print(a, b):
    print(a + b)

result = add_and_print(2, 3)
print("Stored result:", result)
````

## 5. Using mutable default arguments carelessly
This is an advanced but important warning.
Avoid defaults like `items=[]`.


````python
# Problematic pattern:
# def bad_append(value, items=[]):
#     items.append(value)
#     return items
#
# print(bad_append(1))
# print(bad_append(2))
````

## Safer pattern


````python
def safe_append(value, items=None):
    if items is None:
        items = []
    items.append(value)
    return items

print(safe_append(1))
print(safe_append(2))
````

# 18. Best Practices

## 1. Use meaningful function names
Good:
```python
calculate_total()
is_even()
print_report()
```

## 2. Keep functions focused
A function should ideally do one clear task.

## 3. Use docstrings
They improve readability and maintainability.

## 4. Prefer `return` over only printing
Returning values makes functions more reusable.

## 5. Use default parameters carefully
They are useful, but mutable defaults should be avoided.

## 6. Use `*args` and `**kwargs` only when needed
They are powerful, but overusing them can reduce readability.

## 7. Follow argument order rules
A common order is:
- normal parameters
- default parameters
- `*args`
- keyword-only parameters (if any)
- `**kwargs`

**Answer:**  
Yes, as long as the parameter names are correct.

**Answer:**  
Positional arguments must come first.

### 7. What is wrong with mutable default arguments like `items=[]`?
