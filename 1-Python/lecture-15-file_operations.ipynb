## 2) Core Idea: `with open(...) as file`

The most common pattern in Python file handling is:

```python
with open("example.txt", "r") as file:
    content = file.read()
```

### Why use `with`?
Using `with` is preferred because Python automatically closes the file after the block finishes.

### General syntax
```python
with open(filename, mode) as file:
    # do operations
```

Where:
- `filename` = file name or full path
- `mode` = how you want to open the file


## 3) File Modes You Must Know

| Mode | Meaning | Notes |
|---|---|---|
| `"r"` | Read | File must already exist |
| `"w"` | Write | Creates file if not present, **overwrites** if present |
| `"a"` | Append | Creates file if not present, adds content at the end |
| `"w+"` | Read + Write | Creates file if needed, **truncates** existing content |
| `"rb"` | Read binary | Used for binary files |
| `"wb"` | Write binary | Used for writing bytes |

## 4) Reading an Entire Text File

### Concept
Use `.read()` to read the full content at once.

### Example
```python
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

### Important behavior
If the file does **not** exist and you open it using `"r"`, Python raises an error:

```python
FileNotFoundError
```

So read mode is only for files that already exist.


````python
# Reading the whole file
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
````

## 5) Reading a File Line by Line

When you loop through a file object, Python gives one line at a time.

```python
with open("example.txt", "r") as file:
    for line in file:
        print(line)
```

### Why extra blank spaces may appear
Each line often ends with a newline character `\n`, so `print(line)` may look like it adds extra spacing.

### Solution
Use `.strip()` to remove leading/trailing whitespace, especially newline characters:

```python
with open("example.txt", "r") as file:
    for line in file:
        print(line.strip())
```


````python
# Reading line by line
with open("example.txt", "r") as file:
    for line in file:
        print(line.strip())
````

## 6) Writing to a File with `"w"`

### Concept
Write mode is used to put content into a file.

```python
with open("example.txt", "w") as file:
    file.write("Hello World\n")
    file.write("This is a new line\n")
```

### Very important warning
If the file already has content, `"w"` will **erase the old content first**.

This is called **overwriting** or **truncating** the file.


````python
# Writing to a file (overwrites existing content)
with open("example.txt", "w") as file:
    file.write("Hello World\n")
    file.write("This is a new line\n")
````

## 7) Appending to a File with `"a"`

### Concept
Append mode adds new content to the **end** of the file without deleting old data.

```python
with open("example.txt", "a") as file:
    file.write("Append operation taking place\n")
```

### Use this when
- You want to keep existing data
- You want to add logs, results, notes, or extra lines


````python
# Appending without overwriting
with open("example.txt", "a") as file:
    file.write("Append operation taking place\n")
````

## 8) Writing Multiple Lines with `writelines()`

If you already have a list of strings, you can write them all together.

```python
lines = [
    "This is my first line\n",
    "This is my second line\n",
    "This is my third line\n"
]

with open("example.txt", "a") as file:
    file.writelines(lines)
```

### Note
`writelines()` does **not** automatically insert newline characters.  
You must include `\n` yourself inside each string if you want separate lines.


````python
# Writing multiple lines
lines = [
    "This is my first line\n",
    "This is my second line\n",
    "This is my third line\n"
]

with open("example.txt", "a") as file:
    file.writelines(lines)
````

## 10) Copying Content from One File to Another

This is a very common real-world task.

```python

    content = source_file.read()

```

### Important behavior

- If it exists, its previous content is overwritten.


````python
# Copy text from source file to destination file
with open("example.txt", "r") as source_file:
    content = source_file.read()

with open("destination.txt", "w") as destination_file:
    destination_file.write(content)
````

## 11) Counting Lines, Words, and Characters

This is a classic practice question.

def count_file_stats(file_path):
    with open(file_path, "r") as file:

## 12) Reading and Writing in the Same File with `"w+"`

### What `"w+"` does
- Opens a file for both **reading and writing**
- Creates the file if it does not exist
- **Overwrites/truncates** the file if it already exists

### Example
```python
with open("example.txt", "w+") as file:
    file.write("Hello World\n")
    file.write("This is a new text line\n")
    file.seek(0)
    content = file.read()
    print(content)
```

### Why `seek(0)` is important
After writing, the file cursor is at the end of the file.  
If you immediately call `.read()`, Python reads from the current position onward — usually nothing.

So you must move the cursor back to the beginning:

```python
file.seek(0)
```

### Key idea
`seek(0)` resets the file pointer to the start of the file.


````python
with open("example.txt", "w+") as file:
    file.write("Hello World\n")
    file.write("This is a new text line\n")

    file.seek(0)  # Move cursor to the beginning
    content = file.read()
    print(content)
````

## 16) Joining Paths Safely

Hardcoding paths is risky because path separators differ across systems.

- Windows often uses `\`
- Linux / macOS use `/`

So instead of manual joining, use:

```python
os.path.join(folder_name, file_name)
```

### Example
```python
folder_name = "folder"
file_name = "file.txt"

full_path = os.path.join(folder_name, file_name)
print(full_path)
```

This makes your code more portable.


````python
folder_name = "folder"
file_name = "file.txt"

full_path = os.path.join(folder_name, file_name)
print(full_path)
````

## 22) Common Mistakes to Avoid

1. **Using `"r"` on a file that does not exist**
   - Causes `FileNotFoundError`

2. **Using `"w"` when you wanted to preserve old content**
   - `"w"` overwrites everything

3. **Forgetting `\n` while writing multiple lines**
   - Lines may appear combined

4. **Using `writelines()` without newline characters**
   - Output becomes one long line

5. **Reading immediately after writing without `seek(0)`**
   - You may get an empty result

6. **Hardcoding path separators**
   - Prefer `os.path.join()`

7. **Confusing text data and bytes**
   - Text files use strings
   - Binary files use bytes

## 26) Tricky / Interview Concepts

### Q1. Why is `with open(...)` preferred over plain `open()`?
**Answer:** Because `with` automatically closes the file, even if an exception occurs.

---

### Q2. What happens if you call `read()` twice?
**Answer:** The first `read()` moves the cursor to the end. The second `read()` usually returns an empty string unless you reset the cursor with `seek(0)`.

---

### Q3. What is a file pointer / cursor?
**Answer:** It is the current position inside the file from where reading or writing happens.

---

### Q4. Why can `print(line)` show extra blank lines?
**Answer:** Because each line already contains a newline character, and `print()` also moves to the next line.

---

### Q5. Does `writelines()` add newline characters automatically?
**Answer:** No. You must manually include `\n`.

---

### Q6. What is the risk of `"w+"` mode?
**Answer:** It truncates the file first, so existing content is lost.

---

### Q7. When should you use binary mode?
**Answer:** For non-text data such as images, audio, PDFs, pickled objects, or custom binary formats.

---

### Q8. What is the difference between relative and absolute paths?
**Answer:** Relative paths depend on the current working directory; absolute paths point to the full exact location.

---

### Q9. Why use `os.path.join()` instead of manually writing paths?
**Answer:** It creates paths safely across operating systems.

---

### Q10. How would you safely create a directory without crashing if it already exists?
**Answer:**
```python
import os

if not os.path.exists("data"):
    os.mkdir("data")
```

Or more modern:
```python
os.makedirs("data", exist_ok=True)
```
