## 1) Big Picture

This lesson covers **decorators**, an advanced and very useful concept in Python.

Before understanding decorators properly, the transcript first explains two supporting ideas:

- **Function copy**
- **Closures**

That sequence is helpful because decorators are built on the idea that:
- functions are first-class objects in Python
- functions can be passed around
- functions can return other functions
- inner functions can “remember” outer variables

Once those concepts are clear, decorators become much easier to understand.


## 2) What Is a Decorator?

A **decorator** is a powerful Python feature that allows you to modify or extend the behavior of a function or method **without changing its original source code**.

### Simple meaning
A decorator “wraps” another function and adds extra behavior before, after, or around it.

### Why decorators are useful
They are commonly used for:
- logging
- authentication
- timing
- access control
- retries
- validation
- caching
- framework features (Flask, Django, FastAPI, etc.)


## 3) Important Foundation: Functions Are First-Class Objects

In Python, functions are **first-class objects**.

That means you can:
- assign them to variables
- pass them as arguments
- return them from other functions
- store them in data structures

This is the foundation of both closures and decorators.


## 4) Function Copy / Function Assignment

The transcript calls this **function copy**.

### Example
```python
def welcome():
    return "Welcome to the Advanced Python course"

well = welcome
```

Now `well` refers to the same function object as `welcome`.

That means:
```python
print(well())
```

will still work exactly like:
```python
print(welcome())
```


````python
def welcome():
    return "Welcome to the Advanced Python course"

well = welcome

print(welcome())
print(well())
print(type(well))
````

## 5) What Happens If the Original Name Is Deleted?

If a function is assigned to another variable, that new variable still points to the function object.

### Example
```python
def welcome():
    return "Welcome to the Advanced Python course"

well = welcome
del welcome

print(well())
```

This still works because `well` still holds a reference to the function.


````python
def welcome():
    return "Welcome to the Advanced Python course"

well = welcome
del welcome

print(well())
````

## 6) Why Function Assignment Matters for Decorators

Decorators rely on the fact that functions can be:
- stored in variables
- passed into other functions
- returned from other functions

If functions were not first-class objects, decorators would not work this way.


## 7) What Is a Closure?

A **closure** happens when:
- a function is defined inside another function
- the inner function uses variables from the outer function
- the inner function remembers those outer variables even after the outer function finishes

### Simple idea
A closure is usually an **inner function plus remembered outer state**.


## 8) Basic Closure Example

```python
def main_welcome(message):
    def sub_welcome():
        print("Welcome to the Advanced Python course")
        print(message)
        print("Please learn these concepts properly")
    return sub_welcome
```

Here:
- `sub_welcome()` is inside `main_welcome()`
- `sub_welcome()` uses `message`, which belongs to the outer function
- returning `sub_welcome` creates a closure


````python
def main_welcome(message):
    def sub_welcome():
        print("Welcome to the Advanced Python course")
        print(message)
        print("Please learn these concepts properly")
    return sub_welcome

fn = main_welcome("Welcome everyone")
print(fn)
fn()
````

## 9) Why This Is Called a Closure

The key part is that the inner function remembers the variable `message` from the outer scope.

Even though `message` belongs to `main_welcome`, the returned `sub_welcome()` can still use it later.

That “remembering” behavior is what makes it a closure.


## 10) Returning a Function, Not Calling It Immediately

Notice this line:

```python
return sub_welcome
```

This returns the **function itself**, not the result of calling it.

### Difference
```python
return sub_welcome
```
means “return the function object”

while

```python
return sub_welcome()
```
means “execute it now and return its result”

For decorators, returning the function object is extremely important.


## 11) Closures Can Also Accept Functions as Arguments

The transcript then changes the outer function so it accepts another function as input.

### Example idea
```python
def main_welcome(func):
    def sub_welcome():
        print("Welcome to the Advanced Python course")
        func("Welcome everyone to this tutorial")
        print("Please learn these concepts properly")

```

Now the outer function is receiving another function and using it inside the inner function.


````python
def main_welcome(func):
    def sub_welcome():
        print("Welcome to the Advanced Python course")
        func("Welcome everyone to this tutorial")
        print("Please learn these concepts properly")
    return sub_welcome

wrapped = main_welcome(print)
wrapped()
````

## 12) Functions Can Be Passed as Arguments

So when the inner function runs:

## 14) How Closures Lead to Decorators

Decorators are built from the same structure:

1. outer function receives a function
2. outer function defines an inner wrapper function
3. wrapper adds extra behavior
4. outer function returns the wrapper

That is the standard decorator pattern.


## 15) Manual Decorator Pattern

### Example
```python
def main_welcome(func):
    def sub_welcome():
        print("Welcome to the Advanced Python course")
        func()
        print("Please learn these concepts properly")
    return sub_welcome
```

This is essentially a decorator.

If we then define:
```python
def course_introduction():
    print("This is an advanced Python course")
```

we can manually do:
```python
decorated = main_welcome(course_introduction)
decorated()
```


````python
def main_welcome(func):
    def sub_welcome():
        print("Welcome to the Advanced Python course")
        func()
        print("Please learn these concepts properly")
    return sub_welcome

def course_introduction():
    print("This is an advanced Python course")

decorated = main_welcome(course_introduction)
decorated()
````

## 16) The `@decorator_name` Syntax

Instead of wrapping functions manually, Python gives us a cleaner syntax:

```python
@main_welcome
def course_introduction():
    print("This is an advanced Python course")
```

This is equivalent to:

```python
def course_introduction():
    print("This is an advanced Python course")

course_introduction = main_welcome(course_introduction)
```

That is the real meaning of decorator syntax.


````python
def main_welcome(func):
    def sub_welcome():
        print("Welcome to the Advanced Python course")
        func()
        print("Please learn these concepts properly")
    return sub_welcome

@main_welcome
def course_introduction():
    print("This is an advanced Python course")

course_introduction()
````

## 17) Most Important Decorator Idea

Decorators let you add extra behavior **without modifying the original function code**.

### Example
Original function:
```python
def course_introduction():
    print("This is an advanced Python course")
```

Decorator adds:
- text before
- text after
- wrapping behavior

The original function body stays unchanged.


## 18) A Classic Simple Decorator Example

The transcript gives a cleaner general-purpose example:

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper
```

This is the most standard beginner decorator pattern.


````python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello")

say_hello()
````

## 19) Decorators with Function Arguments

A normal decorator like the previous one works only for functions with no arguments.

If the decorated function accepts parameters, the wrapper must also accept them.

The transcript introduces:

- `*args`
- `**kwargs`

These allow the decorator to handle any number of positional and keyword arguments.


## 20) Decorator with `*args` and `**kwargs`

### Example
```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before function call")
        result = func(*args, **kwargs)
        print("After function call")
        return result
    return wrapper
```

This is the standard flexible decorator form.


````python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Before function call")
        result = func(*args, **kwargs)
        print("After function call")
        return result
    return wrapper

@my_decorator
def greet(name):
    print(f"Hello, {name}")

greet("Krish")
````

## 21) Decorator That Takes Its Own Arguments

The transcript then shows a more advanced case:

```python
@repeat(3)
def say_hello():
    print("Hello")
```

This means the decorator itself is parameterized.

That requires **three layers**:

1. outer function that receives decorator arguments (`n`)
2. decorator function that receives the function
3. wrapper function that runs the actual call


## 22) Example: Repeat Decorator

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

### Usage
```python
@repeat(3)
def say_hello():
    print("Hello")
```

Now calling `say_hello()` prints `Hello` three times.


````python
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(3)
def say_hello():
    print("Hello")

say_hello()
````

## 23) Understanding the Three Layers in `repeat`

### Layer 1: outer function
```python
def repeat(n):
```
Receives decorator configuration like `3`

### Layer 2: actual decorator
```python
def decorator(func):
```
Receives the function being decorated

### Layer 3: wrapper
```python
def wrapper(*args, **kwargs):
```
Runs when the decorated function is actually called

This pattern is very common in advanced Python.


## 24) Why `*args` and `**kwargs` Matter

They make decorators reusable.

Without them, your decorator may only work on functions with one exact signature.

With them, it can decorate many different functions.

### Example
```python
def wrapper(*args, **kwargs):
    return func(*args, **kwargs)
```

This passes through all arguments safely.


## 25) Important Correction: Preserve Return Values

One thing beginners often miss is that a wrapper should usually return the original function's result.

### Good pattern
```python
def wrapper(*args, **kwargs):
    result = func(*args, **kwargs)
    return result
```

If you forget to return the result, decorated functions that should return values may stop working correctly.


````python
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"Calling function: {func.__name__}")
        result = func(*args, **kwargs)
        print(f"Function {func.__name__} finished")
        return result
    return wrapper

@logger
def add(a, b):
    return a + b

print(add(10, 20))
````

## 26) Important Correction: Use `functools.wraps`

A common issue with decorators is that they can hide the metadata of the original function:
- function name
- docstring
- module info

To preserve metadata, use:

```python
from functools import wraps
```

### Example
```python
from functools import wraps

def my_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)
    return wrapper
```

This is the professional best practice.


````python
from functools import wraps

def my_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print("Before function call")
        result = func(*args, **kwargs)
        print("After function call")
        return result
    return wrapper

@my_decorator
def greet(name):
    """Greets a user."""
    print(f"Hello, {name}")

print(greet.__name__)
print(greet.__doc__)
greet("Lucy")
````

## 27) Common Real-World Uses of Decorators

Decorators are heavily used in real applications.

### Common examples
- logging function calls
- measuring execution time
- authorization checks

- route registration in web frameworks
- caching expensive computations
- input validation

Frameworks like Flask and Django use decorators a lot.


## 28) Common Mistakes to Avoid

1. **Forgetting to return the wrapper**
   - then the decorator does not work

2. **Forgetting to call the wrapped function**
   - wrapper runs, original function never executes

3. **Not using `*args, **kwargs`**
   - decorator becomes too limited

4. **Forgetting to return the original function’s result**
   - breaks functions that return values

5. **Confusing `@decorator` with `@decorator()`**
   - the second form is for parameterized decorators

6. **Not using `functools.wraps`**
   - loses original function metadata


## 29) Important Corrections from the Transcript

The transcript is conceptually helpful, but here are some cleaner explanations:

1. **Decorator syntax uses `@`, not an “at the rate keyword”**
   - It is commonly read as “at”

2. **Closures are not just “a function inside a function”**
   - More accurately: a closure is an inner function that remembers variables from its enclosing scope

3. **Function copy is really function assignment / aliasing**
   - the same function object is referenced by another name

4. **Passing `len` requires passing suitable data**
   - `len` expects a sized object like a list or string, not an integer

5. **Decorator does not magically run everything**
   - it replaces the decorated function name with the wrapper returned by the decorator

## 33) Tricky / Interview Concepts

### Q1. What exactly does a decorator return?
**Answer:**  
Usually a wrapper function that replaces the original function reference.

---

### Q2. Are decorators only for functions?
**Answer:**  
No. Decorators can also be used with methods and classes.

---

### Q3. Why are closures important in decorators?
**Answer:**  
Because the wrapper function often “remembers” the original function and any decorator arguments.

---

### Q4. What is metadata loss in decorators?
**Answer:**  
Without `functools.wraps`, the decorated function may lose its original `__name__`, docstring, and other metadata.

---

### Q5. Why do frameworks use decorators heavily?
**Answer:**  
Because decorators are a clean way to register routes, apply access control, validation, or configuration.

---

### Q6. Can multiple decorators be stacked?
**Answer:**  
Yes. Python applies them from the bottom upward.

---

### Q7. What is the difference between a simple decorator and a decorator factory?
**Answer:**  
A simple decorator directly accepts a function. A decorator factory first accepts configuration arguments and then returns a decorator.

---

### Q8. Why is `wrapper(*args, **kwargs)` considered best practice?
**Answer:**  
Because it keeps the decorator compatible with many different functions.

---

### Q9. What is the mental model of decorator syntax?
**Answer:**  
It is just function reassignment after wrapping.

---

### Q10. Is a decorator modifying the original function source?
**Answer:**  
Usually no. It wraps or replaces the callable reference rather than editing the body itself.

## 35) Best Practices

1. Use decorators when behavior should be reusable across many functions
2. Use `*args, **kwargs` in wrappers unless you have a very specific reason not to
3. Return the original function’s result
4. Use `functools.wraps`
5. Keep decorators focused and readable
6. Use parameterized decorators when behavior needs configuration
7. Avoid overusing decorators when plain functions would be clearer
