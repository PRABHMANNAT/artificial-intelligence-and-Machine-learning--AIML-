# 1. Introduction to Loops

A loop allows you to execute a block of code multiple times.

Loops are useful when:
- repeating a task
- processing collections of data
- counting
- searching
- generating patterns
- doing repeated calculations

## Main loop types in Python
- `for` loop
- `while` loop

## Loop-related control statements
- `break`
- `continue`
- `pass`


# 2. The `for` Loop

A `for` loop is used to iterate over an **iterable**.

## What is an iterable?
An iterable is something you can loop through, such as:
- a `range`
- a string
- a list
- a tuple
- a set
- a dictionary

## Basic syntax
```python
for item in iterable:
    # code block
```


## Example: simple `for` loop with `range()`


````python
for i in range(5):
    print(i)
````

### Explanation
`range(5)` gives:
```python
0, 1, 2, 3, 4
```

So the loop prints those values one by one.


# 3. Understanding `range()`

`range()` is commonly used with `for` loops.

## Common forms of `range()`

### 1. `range(stop)`
```python
range(5)
```
Produces:
```python
0, 1, 2, 3, 4
```

### 2. `range(start, stop)`
```python
range(1, 6)
```
Produces:
```python
1, 2, 3, 4, 5
```

### 3. `range(start, stop, step)`
```python
range(1, 10, 2)
```
Produces:
```python
1, 3, 5, 7, 9
```


````python
print(list(range(5)))
print(list(range(1, 6)))
print(list(range(1, 10, 2)))
````

## Important rule
The `stop` value is always excluded.

That is why:
```python
range(1, 6)
```
gives:
```python
1, 2, 3, 4, 5
```
not 6.


## Using a negative step
A negative step lets you count backward.


````python
print(list(range(10, 0, -1)))
print(list(range(10, 0, -2)))
````

### Explanation
- `range(10, 0, -1)` gives `10` down to `1`
- `0` is excluded because the stop value is not included

# 6. The `while` Loop

A `while` loop keeps running **as long as a condition is true**.

## Basic syntax
```python
while condition:
    # code block
```

The loop stops when the condition becomes false.


````python
count = 0

while count < 5:
    print(count)
    count += 1
````

## Explanation
- Start with `count = 0`
- print the count
- increase it by 1
- repeat while `count < 5`

So the output is:
```python
0, 1, 2, 3, 4
```


## Important warning
In a `while` loop, you must usually update the loop variable.
Otherwise, the condition may never become false, causing an **infinite loop**.


### Incorrect example
```python
count = 0
while count < 5:
    print(count)
```

This keeps printing forever because `count` never changes.


# 7. Another `while` Example


````python
count = 0

while count % 2 == 0:
    print("Count =", count)
    count += 1
````

### What happens here?
- Initially `count = 0`
- `0 % 2 == 0` is true, so the loop runs once
- then `count` becomes `1`
- `1 % 2 == 0` is false, so the loop stops


# 8. `break` Statement

The `break` statement immediately exits the loop.

## Basic idea
When `break` runs:
- the current loop stops immediately
- control moves to the first statement after the loop


````python
for i in range(10):
    if i == 5:
        break
    print(i)
````

## Output explanation
The output is:
```python
0
1
2
3
4
```

When `i` becomes `5`, the `break` statement runs **before** `print(i)`, so 5 is not printed.


## Common use cases of `break`
- stop searching when target is found
- exit loop when user enters a stop command
- terminate early on some special condition


# 9. `continue` Statement

The `continue` statement skips the current iteration and moves to the next one.


````python
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
````

## Explanation
If `i` is even, `continue` skips the rest of that iteration.

So only odd numbers are printed:
```python
1, 3, 5, 7, 9
```


## Difference between `break` and `continue`
- `break` → stop the loop completely
- `continue` → skip just the current iteration


# 10. `pass` Statement

`pass` is a null statement — it does nothing.

It is used as a placeholder when Python expects a statement but you do not want to write real logic yet.


````python
for i in range(5):
    if i == 3:
        pass
    print(i)
````

## Explanation
Even when `i == 3`, `pass` does nothing.
So the loop still prints every value:
```python
0, 1, 2, 3, 4
```

This is different from `continue`, which would skip the print for that iteration.


````python
for i in range(5):
    if i == 3:
        continue
    print(i)
````

### Compare carefully
In the second example, `3` is not printed, because `continue` skips the rest of that loop iteration.


## Common use cases of `pass`
- placeholder in empty functions
- placeholder in empty classes
- placeholder while writing code step by step


````python
def future_function():
    pass

class FutureClass:
    pass

print("Defined placeholder function and class successfully.")
````

# 11. Nested Loops

A nested loop means a loop inside another loop.

## General form
```python
for i in ...:
    for j in ...:
        ...
```

The inner loop completes all its iterations for each one iteration of the outer loop.


````python
for i in range(3):
    for j in range(2):
        print(f"i = {i}, j = {j}")
````

## Explanation
For each value of `i`, the entire inner loop over `j` runs.

So the sequence is:
- `i = 0`, `j = 0`
- `i = 0`, `j = 1`
- `i = 1`, `j = 0`
- `i = 1`, `j = 1`
- `i = 2`, `j = 0`
- `i = 2`, `j = 1`


# 12. Practical Example: Sum of First `n` Natural Numbers Using `while`


````python
n = 10
total = 0
count = 1

while count <= n:
    total += count
    count += 1

print(f"Sum of first {n} natural numbers = {total}")
````

### Explanation
This computes:
```python
1 + 2 + 3 + ... + 10
```
which gives `55`.


# 13. Practical Example: Sum of First `n` Natural Numbers Using `for`


````python
n = 10
total = 0

for i in range(1, n + 1):
    total += i

print(f"Sum of first {n} natural numbers = {total}")
````

## Why `range(1, n + 1)`?
Because the stop value is excluded, so we use `n + 1` to include `n`.


# 14. Practical Example: Prime Numbers Between 1 and 100

A prime number is a number greater than 1 that has exactly two factors:
- 1
- itself


````python
for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num)
````

## Explanation of `for-else`
In Python, a `for` loop can have an `else` block.

The `else` block runs **only if the loop finishes normally** without hitting `break`.

In the prime-number example:
- if a divisor is found → `break`
- if no divisor is found → `else` runs, so the number is prime


# 15. Another Prime Check Example as a Function


````python
def is_prime(num: int) -> bool:
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

for n in [2, 3, 4, 5, 11, 12, 13]:
    print(n, "->", is_prime(n))
````

# 16. Practical Example: Print a Multiplication Table


````python
number = 5

for i in range(1, 11):
    print(f"{number} x {i} = {number * i}")
````

# 17. Practical Example: Pattern with Nested Loops


````python
for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()
````

# 18. Common Errors in Loops

## 1. Infinite loop
Example:
```python
count = 0
while count < 5:
    print(count)
```

Problem:
- `count` is never updated
- condition stays true forever


## 2. Off-by-one error
Example:
```python
for i in range(1, 10):
    print(i)
```

This prints 1 to 9, not 1 to 10.

To include 10:
```python
for i in range(1, 11):
    print(i)
```


## 3. Misunderstanding `break`
`break` exits the loop completely — it does not just skip one iteration.


## 4. Misunderstanding `continue`
`continue` skips the rest of the current iteration — it does not stop the loop entirely.


## 5. Misunderstanding `pass`
`pass` does nothing. It does not skip printing, and it does not stop the loop.


## 6. Wrong indentation
Incorrect indentation changes loop behavior or raises errors.


# 19. Best Practices

## 1. Use meaningful variable names
Prefer:
```python
for number in range(5):
```
instead of unclear names unless short loop variables are natural.

## 2. Avoid deeply nested loops unless needed
They can become hard to read and may be slower.

## 3. Be careful with `while` loops
Always ensure the loop condition will eventually become false.

## 4. Use `break` and `continue` only when they improve clarity
Overusing them can make flow harder to understand.

## 5. Know your range boundaries
Remember: stop value is excluded.

## 6. Prefer `for` when iterating over a collection
Use `while` mainly when the number of iterations depends on a changing condition.

# 23. Interview Concepts and Model Answers

### 1. What is the difference between `for` and `while` loops in Python?
**Model answer:**  
A `for` loop is typically used to iterate over an iterable like a list, string, or range. A `while` loop is used when repetition depends on a condition staying true.

### 2. What is `range()` and how does it work?
**Model answer:**  
`range()` generates a sequence of integers. It can take one, two, or three arguments: `range(stop)`, `range(start, stop)`, and `range(start, stop, step)`. The stop value is excluded.

### 3. What is the difference between `break`, `continue`, and `pass`?
**Model answer:**  
`break` exits the loop immediately, `continue` skips the current iteration and moves to the next one, and `pass` does nothing and is used as a placeholder.

### 4. What is an infinite loop?
**Model answer:**  
An infinite loop is a loop that never terminates because its condition never becomes false or it lacks a proper exit mechanism.

### 5. What are nested loops?
**Model answer:**  
Nested loops are loops inside another loop. For each iteration of the outer loop, the inner loop runs completely.

### 6. When would you use a `while` loop instead of a `for` loop?
**Model answer:**  
Use a `while` loop when the number of iterations is not known in advance and depends on a condition or runtime event.

### 7. Can you iterate through a string using a loop?
**Model answer:**  
Yes. A string is iterable, so a `for` loop can iterate character by character.

### 8. What is the purpose of `for-else` in Python?
**Model answer:**  
The `else` block of a `for` loop executes if the loop finishes normally without encountering a `break`.

### 9. What is an off-by-one error?
**Model answer:**  
It is a boundary mistake where a loop runs one time too many or one time too few, often caused by misunderstanding range endpoints.

### 10. How do you write clean loop logic?
**Model answer:**  
Use clear variable names, understand loop boundaries, avoid unnecessary nesting, update variables properly in `while` loops, and use `break`/`continue` only when they improve clarity.
