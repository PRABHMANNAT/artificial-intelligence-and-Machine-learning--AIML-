# 1. Introduction to Abstraction

Abstraction is one of the major pillars of object-oriented programming.

## Simple definition
Abstraction means:
- hiding complex implementation details
- showing only the necessary features of an object

## Real-world intuition
You use many devices through simple controls:
- washing machine buttons
- AC remote
- phone apps
- laptop shutdown button

You see the **interface**, but the internal implementation is hidden.

That is the core idea of abstraction.


# 2. Why Abstraction Matters

Abstraction helps because it:
- reduces complexity
- hides unnecessary internal details
- improves readability
- makes systems easier to use
- encourages cleaner design


# 3. Abstraction in Python

In Python, abstraction is commonly modeled using the `abc` module.

## Important pieces
- `ABC` → Abstract Base Class
- `@abstractmethod` → marks methods that subclasses must implement


## Basic import


````python
from abc import ABC, abstractmethod
````

# 4. Creating an Abstract Base Class

Let us build a clean version of the transcript example.


````python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def drive(self):
        return "The vehicle is used for driving."

    @abstractmethod
    def start_engine(self):
        pass
````

## Explanation
Here:
- `Vehicle` is an abstract base class
- `drive()` is a normal concrete method
- `start_engine()` is an abstract method

This means:
- subclasses can directly use `drive()`
- subclasses **must** implement `start_engine()`


# 5. Concrete Child Class Example

Now let us create a concrete child class that inherits from `Vehicle`.


````python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def drive(self):
        return "The vehicle is used for driving."

    @abstractmethod
    def start_engine(self):
        pass


class Car(Vehicle):
    def start_engine(self):
        return "Car engine started."


def operate_vehicle(vehicle):
    print(vehicle.start_engine())
    print(vehicle.drive())


car = Car()
operate_vehicle(car)
````

## Explanation
The `Car` class:
- inherits from `Vehicle`
- implements `start_engine()`
- automatically gets access to `drive()`

So `Car` becomes a valid concrete class.


# 6. Why Abstract Methods Are Useful

Abstract methods create a **contract**.

They say:
> Any subclass of this abstract class must provide its own implementation of these required methods.

This makes designs more consistent and structured.


## What happens if a subclass does not implement the abstract method?
It cannot be instantiated.


````python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass


class Bike(Vehicle):
    pass

# Uncommenting the next line would raise an error:
# bike = Bike()
````

# 7. Abstraction Example with Multiple Child Classes

This shows how abstraction helps define a common interface for related objects.


````python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def drive(self):
        return "Vehicle is used for driving."

    @abstractmethod
    def start_engine(self):
        pass


class Car(Vehicle):
    def start_engine(self):
        return "Car engine started."


class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle engine started."


class Truck(Vehicle):
    def start_engine(self):
        return "Truck engine started."


vehicles = [Car(), Motorcycle(), Truck()]

for v in vehicles:
    print(v.start_engine())
    print(v.drive())
````

## Key idea
All child classes share the same public interface:
- `start_engine()`
- `drive()`

But each class can implement `start_engine()` differently.

## Abstraction

In practice, abstraction is often implemented using inheritance plus abstract base classes.

# 11. Introduction to Magic Methods

Magic methods are also called **dunder methods**.

“Dunder” means:
- **double underscore**
- method names that start and end with `__`

Examples:
- `__init__`
- `__str__`
- `__repr__`


## Why magic methods matter
Magic methods let you define how your objects behave for built-in operations such as:
- object creation
- printing
- string conversion
- comparisons
- arithmetic
- iteration
and more


# 12. `__init__`: Object Initialization

`__init__` is the constructor-like method used when creating an object.


````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


person = Person("Krish", 34)
print(person.name)
print(person.age)
````

## Explanation
When you write:
```python
person = Person("Krish", 34)
```

Python calls:
```python
__init__(self, "Krish", 34)
```

and initializes the object state.


# 13. Default Object Representation

If you print an object without defining `__str__`, Python shows a default representation.


````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


person = Person("Krish", 34)
print(person)
````

## Explanation
The default output usually looks like:
```python
<__main__.Person object at 0x...>
```

That is the default object representation.


# 14. Overriding `__str__`

`__str__` returns a user-friendly string representation of the object.


````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, {self.age} years old"


person = Person("Krish", 34)
print(person)
````

## Explanation
Now when `print(person)` runs, Python uses `__str__()`.

This makes the object display more readable.


# 15. Overriding `__repr__`

`__repr__` returns a more official / developer-oriented representation.


````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"Person(name={self.name!r}, age={self.age!r})"


person = Person("Krish", 34)
print(repr(person))
````

## Explanation
`repr(person)` explicitly asks for the `__repr__()` representation.

A good `__repr__()` often looks like something useful for debugging.


# 16. Using Both `__str__` and `__repr__`

You can define both in the same class.


````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, {self.age} years old"

    def __repr__(self):
        return f"Person(name={self.name!r}, age={self.age!r})"


person = Person("Krish", 34)

print("print(person):", person)
print("repr(person):", repr(person))
````

## Rule of thumb
- `__str__` → readable for users
- `__repr__` → informative for developers


# 17. Why Magic Methods Are Powerful

Magic methods let you customize how Python interacts with your objects.

This means your objects can feel more natural and more “Pythonic.”


## Examples of other magic methods
- `__len__` → behavior for `len(obj)`
- `__add__` → behavior for `obj1 + obj2`
- `__eq__` → behavior for `obj1 == obj2`
- `__getitem__` → behavior for indexing


# 18. Example: Overriding `__len__`

Let us show one more magic method example.


````python
class BookCollection:
    def __init__(self, books):
        self.books = books

    def __len__(self):
        return len(self.books)


collection = BookCollection(["Python", "ML", "DL"])
print(len(collection))
````

## Explanation
Because `__len__` is defined, Python knows what to do when `len(collection)` is called.


# 19. Example: Overriding `__add__`

Magic methods can also support operator behavior.


````python
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

    def __repr__(self):
        return f"Number({self.value})"


n1 = Number(10)
n2 = Number(20)

print(n1 + n2)
````

## Explanation
Because `__add__` is defined, the `+` operator now works on `Number` objects in a custom way.


# 20. Connecting Abstraction and Magic Methods

These are different ideas, but both improve object-oriented design.

## Abstraction
Defines essential behavior and hides implementation details.

## Magic methods
Control how objects behave with Python’s built-in operations.

Together, they help build clean, expressive, powerful classes.


# 21. Common Errors in Abstraction

## 1. Forgetting to inherit from `ABC`
If you want a formal abstract base class, inherit from `ABC`.

## 2. Forgetting `@abstractmethod`
Without it, the method is not truly enforced as abstract.

## 3. Not implementing required abstract methods
Then the subclass cannot be instantiated.

## 4. Treating abstraction as “just empty functions”
The real point is enforcing a required interface, not emptiness itself.


# 22. Common Errors with Magic Methods

## 1. Using the wrong method name
`__str__` and `__repr__` must be written exactly.

## 2. Returning non-string values from `__str__` or `__repr__`
These methods should return strings.

## 3. Expecting `print(obj)` to use `__repr__` first
Usually `print(obj)` uses `__str__`, while `repr(obj)` uses `__repr__`.

## 4. Overusing magic methods
Use them when they make your object behavior more natural, not just for novelty.


# 23. Best Practices

## For abstraction
- use abstract base classes when you need a strict common interface
- keep abstract methods meaningful
- let child classes implement specialized behavior

## For magic methods
- implement `__str__` for user-friendly display
- implement `__repr__` for debugging clarity
- keep magic method behavior intuitive
- return correct types from magic methods
